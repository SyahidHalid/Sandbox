```python
# Fucntion to do binary mapping

def binary_encoding(df, column, positive_value):
    df = df.copy()
    df[column] = df[column].apply(lambda x: 1 if x == positive_value else 0)
    return df
```

```python
# Apply binary mapping 

credit_card_user_chrun_data = binary_encoding(credit_card_user_chrun_data, column='Attrition_Flag', positive_value='Attrited Customer')

sns.countplot(x="Attrition_Flag", data=credit_card_user_chrun_data)
```

```python
#buat function to preprocess maritial status column

def preprocess_marital_status(data):
    # buat colum dummy cm td
    marital_dummies = pd.get_dummies(data['Marital_Status'])
    
    #add dumy to dataset
    data=pd.concat([data,marital_dummies], axis=1)
    
    #drop original
    data.drop("Marital_Status", axis=1, inplace=True)
    
    return data
```

```python
credit_card_user_chrun_data3 = preprocess_marital_status(credit_card_user_chrun_data2)
```

```python
sns.countplot(x="Attrition_Flag",hue="Income_Category", data=credit_card_user_chrun_data3)
```

```python
def preprocess_income_category(data):
    income_map={'Less than $40K':0,
               "$40K - $60K": 1,
               "$60K - $80K": 2,
               "$80K - $120K": 3,
               "$120K +": 4,}
    
    data['Income_Category'] = data['Income_Category'].apply(lambda x: income_map[x])
    
    return data
```

```python
credit_card_user_chrun_data3 = preprocess_income_category(credit_card_user_chrun_data3)
```

```python
sns.countplot(x="Attrition_Flag",hue="Card_Category", data=credit_card_user_chrun_data3)
```

```python
def preprocess_card_category(data):
    data = pd.get_dummies(data, columns=['Card_Category'])
    
    return data

credit_card_user_chrun_data4 = preprocess_card_category(credit_card_user_chrun_data3)
```

```python
#credit_card_user_chrun_data4.info()
```

# Seperating dependent & independent variables

```python
x = credit_card_user_chrun_data4.drop(['Attrition_Flag'], axis=1)
y = credit_card_user_chrun_data4['Attrition_Flag']
```

```python
x
```

```python
scaler = StandardScaler()
x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
```

```python
x
```

# Training

```python
#Split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=123)
```

```python
x_test
```

```python
print(x_train.shape, '-----> Training Features Shape')
print(x_test.shape,  '-----> Testing Features Shape')
print(y_train.shape, '-----> Training Label Shape')
print(y_test.shape,  '-----> Testing Label Shape')
```

```python
models = [LogisticRegression(),SVC(),DecisionTreeClassifier(),RandomForestClassifier()]

for model in models:
    model.fit(x_train,y_train)
```

```python
model_names = ["Logistic Regression","Support Vactor Machine","Decision Tree","Random Forest"]

for model, name in zip(models, model_names):
    print(name + ": {:.4f}%".format(model.score(x_test, y_test)*100))
```

```python
#using the trained Logistic Regression model to predict

y_pred = models[0].predict(x_test)

y_pred
```

```python
#SVM

y_pred = models[1].predict(x_test)

y_pred
```

```python
#Decision Tree

y_pred = models[2].predict(x_test)

y_pred
```

# Hyperparameter Tuning 

Logistic Regression

```python
logreg = LogisticRegression(random_state=123)

param_grid = {'C':[0.1, 1, 10], 'penalty': ['l1','l2']}

grid_search = GridSearchCV(logreg, param_grid, cv=5)

grid_search.fit(x_train, y_train)

print("Best Hyperparameters:", grid_search.best_params_)
print("")

#make predictions on test data using best estimator
logreg_best = grid_search.best_estimator_
log_prediction = logreg_best.predict(x_test)

print(classification_report(y_test, log_prediction))
print("Confusion:")
cm = confusion_matrix(y_test, log_prediction)
print(cm)

sns.heatmap(cm.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('Acutal Class')
plt.ylabel('Predicted Class')
```

Support Vector Machine

```python
param_grid = {'C':[0.1, 1, 10], 'kernel': ['linear','rbf','sigmoid'], 'gamma': ['scale','auto']}

svm = SVC(random_state=123)

grid_search = GridSearchCV(svm, param_grid, cv=5)

grid_search.fit(x_train, y_train)

print("Best Hyperparameters:", grid_search.best_params_)
print("")

#make predictions on test data using best estimator
svm_best = grid_search.best_estimator_
svm_prediction = svm_best.predict(x_test)

print(classification_report(y_test, svm_prediction))
print("Confusion:")
cm = confusion_matrix(y_test, svm_prediction)
print(cm)

sns.heatmap(cm.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('Acutal Class')
plt.ylabel('Predicted Class')
```

Decision Tree

```python
param_grid = {'max_depth':[3,5,10], 'min_samples_split': [2,5,10], 'min_samples_leaf': [1,2,4]}

dtc = DecisionTreeClassifier(random_state=123)

grid_search = GridSearchCV(dtc, param_grid, cv=5)

grid_search.fit(x_train, y_train)

print("Best Hyperparameters:", grid_search.best_params_)
print("")

#make predictions on test data using best estimator
dtc_best = grid_search.best_estimator_
dtc_prediction = dtc_best.predict(x_test)

print(classification_report(y_test, dtc_prediction))
print("Confusion:")
cm = confusion_matrix(y_test, dtc_prediction)
print(cm)

sns.heatmap(cm.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('Acutal Class')
plt.ylabel('Predicted Class')
```

Random Forest

```python
param_grid = {'n_estimators':[100,300,500], 'max_depth':[5,10,20],
              'min_samples_split': [2,5,10], 'min_samples_leaf': [1,2,4]}

rf = RandomForestClassifier(random_state=123)

grid_search = GridSearchCV(rf, param_grid, cv=5)

grid_search.fit(x_train, y_train)

print("Best Hyperparameters:", grid_search.best_params_)
print("")

#make predictions on test data using best estimator
rf_best = grid_search.best_estimator_
rf_prediction = rf_best.predict(x_test)

print(classification_report(y_test, rf_prediction))
print("Confusion:")
cm = confusion_matrix(y_test, rf_prediction)
print(cm)

sns.heatmap(cm.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('Acutal Class')
plt.ylabel('Predicted Class')
```
