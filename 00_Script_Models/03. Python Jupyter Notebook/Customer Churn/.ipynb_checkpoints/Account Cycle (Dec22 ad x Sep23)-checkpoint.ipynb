```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None)

df = pd.read_excel(r'C:\Users\syahidhalid\OneDrive - Export-Import Bank of Malaysia (EXIM Bank)\Desktop\Work\Customer Churn Prediction\Compiled.xlsx', sheet_name='Sheet1')

Sep23 = df.iloc[np.where((df['Position as At']=='2023-09-30')&(df['Asset Quality Status'].isin(['Active','Active-Watchlist','Impaired'])))]
Aug23 = df.iloc[np.where((df['Position as At']=='2023-08-31')&(df['Asset Quality Status'].isin(['Active','Active-Watchlist','Impaired'])))]
Jul23 = df.iloc[np.where((df['Position as At']=='2023-07-31')&(df['Asset Quality Status'].isin(['Active','Active-Watchlist','Impaired'])))]
Jun23 = df.iloc[np.where((df['Position as At']=='2023-06-30')&(df['Asset Quality Status'].isin(['Active','Active-Watchlist','Impaired'])))]
May23 = df.iloc[np.where((df['Position as At']=='2023-05-31')&(df['Asset Quality Status'].isin(['Active','Active-Watchlist','Impaired'])))]
Dec22 = df.iloc[np.where((df['Position as At']=='2022-12-30')&(df['Asset Quality Status'].isin(['Active','Active-Watchlist','Impaired'])))]
Dec21 = df.iloc[np.where((df['Position as At']=='2021-12-31')&(df['Asset Quality Status'].isin(['Active','Active-Watchlist','Impaired'])))]
Dec20 = df.iloc[np.where((df['Position as At']=='2020-12-31')&(df['Asset Quality Status'].isin(['Active','Active-Watchlist','Impaired'])))]

print(Sep23.shape)
print(Aug23.shape)
print(Jul23.shape)
print(Jun23.shape)
print(May23.shape)
print(Dec22.shape)
print(Dec21.shape)
print(Dec20.shape)
```

# Dec22 - Sep23

```python
# Check Acc yg Dec21 keluar Dec22
```

```python
#yg keluar
combine1 = Dec21.merge(Dec22[['EXIM Account No.']], on='EXIM Account No.', how='outer', indicator=True)
combine1._merge.value_counts()
```

```python
#listing yg keluar
a = combine1.iloc[np.where(combine1._merge=='left_only')].fillna(0)

print(a.shape,
sum(a['Total Loans Outstanding (MYR)']),
sum(a['Cumulative \nProfit Payment/\nInterest Repayment \n(MYR)'])
)
```

```python
b = a.groupby(['CIF Number'])[[]].sum().reset_index()
```

```python
#yg stay
combine2_ = Sep23.merge(b, on='CIF Number', how='inner', indicator=True)
combine2_._merge.value_counts()
```

```python
combine2_.iloc[np.where(combine2_._merge=='both')]['1st Disbursement/Drawdown Date'].value_counts()
```
