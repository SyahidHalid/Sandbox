```python
RDATE1_1 = '31/12/2021'

RDATE = 202112
RDATE1 = 20211231

import numpy as np
import pandas as pd
import pandasql
import sqlite3

pd.set_option("display.max_columns", None) 
pd.set_option("display.max_colwidth", 1000) #huruf dlm column
pd.set_option("display.max_rows", 100)
pd.set_option("display.precision", 2) #2 titik perpuluhan

unit_trust = pd.read_csv(r"T:\MIB Risk Management\Credit Risk Analytics - Historical\02_Data_Source\06_Unit_Trust\04_Dataset\\"+str(RDATE1)+"\\UT_"+str(RDATE1)+".txt", sep = "\t", header = 0, error_bad_lines = False) #, encoding='cp1252'


asb_base = unit_trust.rename(columns = {'M_NPL_DATE':'M_NPL_DATE1'})

asb_base1 = asb_base.iloc[np.where((asb_base.MBB_Entity_Cd=='I')&(asb_base.PRODUCT_DESC.str.contains('ASB')))]

asb_base1.loc[asb_base1.M_NPL_DATE1=='?','M_NPL_DATE1'] = '00/00/0000'

asb_base1['M_NPL_DATE'] = asb_base1.M_NPL_DATE1.str[-4:]+asb_base1.M_NPL_DATE1.str[-7:-5]

asb_base1['OSM'] = asb_base1.Account_Balance_Summary_Amt/1000000

asb_base1['NOA'] = 1

#/* 2 MIA */ 
asb_base1.loc[(asb_base1.M_NPL_FLG!='Y')&(asb_base1.M_Mths_Installment_In_Arrears==2),'OSM_MIA2'] = asb_base1.Account_Balance_Summary_Amt/1000000
asb_base1.loc[(asb_base1.M_NPL_FLG!='Y')&(asb_base1.M_Mths_Installment_In_Arrears==2),'NOA_MIA2'] = 1

#/* 3 MIA */
asb_base1.loc[(asb_base1.M_NPL_FLG!='Y')&(asb_base1.M_Mths_Installment_In_Arrears==3),'OSM_MIA3'] = asb_base1.Account_Balance_Summary_Amt/1000000
asb_base1.loc[(asb_base1.M_NPL_FLG!='Y')&(asb_base1.M_Mths_Installment_In_Arrears==3),'NOA_MIA3'] = 1

#/* IF */
asb_base1.loc[asb_base1.M_NPL_FLG=='Y','OSM_IF'] = asb_base1.Account_Balance_Summary_Amt/1000000
asb_base1.loc[asb_base1.M_NPL_FLG=='Y','NOA_IF'] = 1

#/* NCIF */
asb_base1.loc[asb_base1.M_NPL_DATE==str(RDATE),'OSM_NCIF'] = asb_base1.Account_Balance_Summary_Amt/1000000
asb_base1.loc[asb_base1.M_NPL_DATE==str(RDATE),'NOA_NCIF'] = 1

def TENURE(x):
    if x <= 5:
        return '01.UP TO 5 YEARS'
    elif x > 5 and x <= 10:
        return '02.6 TO 10 YEARS'
    elif x > 5 and x <= 15:
        return '03.11 TO 15 YEARS'
    elif x > 5 and x <= 20:
        return '04.16 TO 20 YEARS'
    elif x > 5 and x <= 25:
        return '05.21 TO 25 YEARS'
    elif x > 5 and x <= 30:
        return '06.26 TO 30 YEARS'
    elif x > 30:
        return '07.ABOVE 30 YEARS'
    else:
        return 'Not Populated'

asb_base1['LOAN_TENURE_GROUP'] = asb_base1.MBB_Loan_Tenure.apply(TENURE)

asb_base1.loc[asb_base1.M_Date_First_Released=='?','M_Date_First_Released'] = asb_base1.M_Date_AA_Approved
asb_base1['MATURITY_DATE'] = asb_base1.M_Date_First_Released.str[-4:].astype(int) + asb_base1.MBB_Loan_Tenure


asb_base1['REMAINING_TENURE'] =  asb_base1['MATURITY_DATE'] - int(str(RDATE)[:4]) 

asb_base1['REM_TENURE_GROUP'] = asb_base1.REMAINING_TENURE.apply(TENURE)

asb_base1['MOB'] = (pd.to_numeric((pd.to_datetime(RDATE1_1, format='%d/%m/%Y') - pd.to_datetime(asb_base1['M_Date_First_Released'], format='%d/%m/%Y')).dt.days, downcast='integer')/30.417) + 1

asb_base1['YEAR_APPROVED'] = asb_base1.M_Date_AA_Approved.str[-4:]

def INCOM(x):
    if x == 'Up to RM999':
        return '01.Below RM3,000'
    elif x == 'RM1,000 to RM1,999':
        return '01.Below RM3,000'
    elif x == 'RM2,000 to RM2,999':
        return '01.Below RM3,000'
    elif x == 'RM3,000 to RM3,999':
        return '02.RM3,000 to RM5,000'
    elif x == 'RM4,000 to RM4,999':
        return '02.RM3,000 to RM5,000'
    elif x == 'RM5,000 to RM5,999':
        return '03.RM5,000 to RM8,000'
    elif x == 'RM6,000 to RM6,999':
        return '03.RM5,000 to RM8,000'
    elif x == 'RM7,000 to RM7,999':
        return '03.RM5,000 to RM8,000'
    elif x == 'RM8,000 to RM8,999':
        return '04.RM8,000 to RM15,000'
    elif x == 'RM9,000 to RM9,999':
        return '04.RM8,000 to RM15,000'
    elif x == 'RM10,000 to RM10,999':
        return '04.RM8,000 to RM15,000'
    elif x == 'RM11,000 to RM14,999':
        return '04.RM8,000 to RM15,000'
    elif x == 'RM15,000 to RM20,000':
        return '05.RM15,000 to RM20,000' 
    elif x == 'Above RM20,000':
        return '06.Above RM20,000'
    else:
        return 'Not Populated'
                
asb_base1['INCOME_GROUP_NEW'] = asb_base1['Income_Group'].apply(INCOM)

def CUSTSEG(x):
    if x == 'MASS':
        return '01.Mass'
    elif x == 'EMERGING AFFLUENT':
        return '02.Emerging Affluent'
    elif x == 'AFFLUENT':
        return '03.Affluent'
    elif x == 'HIGH NETWORTH':
        return '04.High Networth'
    else:
        return '05.Not Populated'

asb_base1['SEGMENT_GROUP'] = asb_base1['CUSTOMER_SEGMENT'].apply(CUSTSEG)

asb_base1.loc[asb_base1.Age_Group == "Not Populated", 'AGE_GROUP_NEW'] = "09.Not Populated"
asb_base1.loc[asb_base1.Age_Group == "<25 years", 'AGE_GROUP_NEW'] = "01.Up to 24 years"
asb_base1.loc[asb_base1.Age_Group == ">25 to <30 years", 'AGE_GROUP_NEW'] = "02.25 to 29 Years Old"
asb_base1.loc[asb_base1.Age_Group == "30 to <35 years", 'AGE_GROUP_NEW'] = "03.30 to 34 Years Old"
asb_base1.loc[asb_base1.Age_Group == "35 to <40 years", 'AGE_GROUP_NEW'] = "04.35 to 39 Years Old"
asb_base1.loc[asb_base1.Age_Group == "40 to <45 years", 'AGE_GROUP_NEW'] = "05.40 to 44 Years Old"
asb_base1.loc[asb_base1.Age_Group == "45 to <50 years", 'AGE_GROUP_NEW'] = "06.45 to 49 Years Old"
asb_base1.loc[asb_base1.Age_Group == "50 to <55 years", 'AGE_GROUP_NEW'] = "07.50 to 54 Years Old"
asb_base1.loc[asb_base1.Age_Group == "55 years and above", 'AGE_GROUP_NEW'] = "08.55 Years & Above"

asb_base1.loc[asb_base1.Region ==  "JOHOR/MELAKA", 'REGION_GROUP'] = "01.JOHOR/MELAKA"
asb_base1.loc[asb_base1.Region ==  "PENANG/KEDAH/PERLIS", 'REGION_GROUP'] = "02.PENANG/KEDAH/PERLIS"
asb_base1.loc[asb_base1.Region ==  "PAHANG/KELANTAN/TERENGGANU", 'REGION_GROUP'] = "03.PAHANG/KELANTAN/TERENGGANU"
asb_base1.loc[asb_base1.Region ==  "SELANGOR/NS", 'REGION_GROUP'] = "04.SELANGOR/NS"
asb_base1.loc[asb_base1.Region ==  "PERAK", 'REGION_GROUP'] = "05.PERAK"
asb_base1.loc[asb_base1.Region ==  "SABAH", 'REGION_GROUP'] = "06.SABAH"
asb_base1.loc[asb_base1.Region ==  "SARAWAK", 'REGION_GROUP'] = "07.SARAWAK"
asb_base1.loc[asb_base1.Region ==  "FEDERAL TERRITORY", 'REGION_GROUP'] = "08.FEDERAL TERRITORY"
asb_base1.loc[asb_base1.Branch ==  "IBS DONGGONGON", 'REGION_GROUP'] = "06.SABAH"
asb_base1.loc[asb_base1.Branch.str.contains("IBS MENARA FELDA, PLATINUM PARK SC"), 'REGION_GROUP'] = "08.FEDERAL TERRITORY"
asb_base1.loc[asb_base1.Branch.str.contains("IBS MATANG JAYA"), 'REGION_GROUP'] = "07.SARAWAK"
asb_base1.loc[asb_base1.Branch.str.contains("IBS ONE CITY USJ"), 'REGION_GROUP'] = "04.SELANGOR/NS"
asb_base1.loc[asb_base1.Branch.str.contains("IBS AEON MALL SHAH ALAM"), 'REGION_GROUP'] = "04.SELANGOR/NS"

asb_base1.loc[asb_base1.RACE_DESC == "CHINESE", 'RACE'] = "01. CHINESE"
asb_base1.loc[asb_base1.RACE_DESC == "INDIAN", 'RACE'] = "02. INDIAN"
asb_base1.loc[asb_base1.RACE_DESC == "MALAY", 'RACE'] = "03. MALAY"
asb_base1.loc[asb_base1.RACE_DESC == "NATIVE", 'RACE'] = "04. NATIVE"
asb_base1.loc[asb_base1.RACE_DESC.isin(["?","OTHERS"]), 'RACE'] = "05. OTHERS"

asb_base1.fillna(0, inplace = True)

```

```python
#asb_base1.OSM = asb_base1.OSM.str.strip()
asb_base1.loc[asb_base1.OSM=='?','OSM'] = 0
asb_base1.OSM = asb_base1.OSM.astype(float)

#asb_base1.EXPECTED_LOSS = asb_base1.EXPECTED_LOSS.str.strip()
asb_base1.loc[asb_base1.EL=='?','EL'] = 0
asb_base1.EL = asb_base1.EL.astype(float)

#asb_base1.EAD = asb_base1.EAD.str.strip()
asb_base1.loc[asb_base1.EAD=='?','EAD'] = 0
asb_base1.EAD = asb_base1.EAD.astype(float)

#asb_base1.RWA = asb_base1.RWA.str.strip()
asb_base1.loc[asb_base1.RWA=='?','RWA'] = 0
asb_base1.RWA = asb_base1.RWA.astype(float)


RISK = asb_base1.fillna(0).groupby(['RISK_GRADE_VAL'])[['OSM','EL','EAD','RWA']].sum().reset_index()
```

```python
#-------------------------------------------SORTBY(EXCEL)

writer2 = pd.ExcelWriter(r'T:\MIB Risk Management\Credit Risk Analytics - Historical\01_Reporting\07_Yearly_Reporting\Budget\BudgetFY21\working - SH\UT_CFS'+str(RDATE)+'(python).xlsx',engine='xlsxwriter')
    

asb_base1.to_excel(writer2, sheet_name='Base', index = False)
RISK.to_excel(writer2, sheet_name='UT', index = False)


writer2.save()
```
