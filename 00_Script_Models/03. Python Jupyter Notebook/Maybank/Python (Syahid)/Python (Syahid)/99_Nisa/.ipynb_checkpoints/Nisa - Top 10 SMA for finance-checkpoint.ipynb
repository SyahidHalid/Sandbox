```python
import numpy as np
import pandas as pd
import pandasql
import sqlite3

pd.set_option("display.max_columns", None) 
pd.set_option("display.max_colwidth", 1000) #huruf dlm column
pd.set_option("display.max_rows", 100)
pd.set_option("display.precision", 2) #2 titik perpuluhan


date = '20220930'

BB = pd.read_excel (r"T:\MIB Risk Management\Credit Risk Analytics - Historical\01_Reporting\01_Monthly_Reporting\10_Credit_Risk_Profile\\"+date[:4]+"\\"+date[:6]+"\\CRP 20220930 -working.xlsx", sheet_name='BB')
CB = pd.read_excel (r"T:\MIB Risk Management\Credit Risk Analytics - Historical\01_Reporting\01_Monthly_Reporting\10_Credit_Risk_Profile\\"+date[:4]+"\\"+date[:6]+"\\CRP 20220930 -working.xlsx", sheet_name='CB')

MasterCB_FRR = pd.read_excel (r'T:\MIB Risk Management\Credit Risk Analytics - Historical\02_Data_Source\11_CRRS\\'+date[:4]+'\\'+date+'\Report_Corp_CRDM_'+date+'.xlsx')


TF='0922'

#BB
SMA_Action_Plan = pd.read_excel (r'T:\MIB Risk Management\Credit Risk Analytics - Historical\02_Data_Source\02_Business_Financing\02_Asset_Quality\\'+date+'\TF'+TF+'-with remarks.xlsx', sheet_name = 'SMA09',header=5)

#CB
CB_SMA_Action_Plan = pd.read_excel (r'T:\MIB Risk Management\Credit Risk Analytics - Historical\02_Data_Source\01_Corporate_Financing\02_Asset_Quality\\'+date+'\CB SMA '+date[:4]+'-'+date[4:6]+'.xlsx', sheet_name = 'SMA',header=7)


date1 = '20220831'
date2 = '20220731'
date3 = '20220630'
date4 = '20220531'
date5 = '20220430'

Rating = pd.read_excel (r"T:\MIB Risk Management\Credit Risk Analytics - Historical\01_Reporting\01_Monthly_Reporting\18_Rating_CBBBRSME\BRR_CBBBRSME_"+date[:6]+".xlsx", sheet_name='Sheet1',header=0)
Rating1 = pd.read_excel (r"T:\MIB Risk Management\Credit Risk Analytics - Historical\01_Reporting\01_Monthly_Reporting\18_Rating_CBBBRSME\BRR_CBBBRSME_"+date1[:6]+".xlsx", sheet_name='Sheet1',header=0)
Rating2 = pd.read_excel (r"T:\MIB Risk Management\Credit Risk Analytics - Historical\01_Reporting\01_Monthly_Reporting\18_Rating_CBBBRSME\BRR_CBBBRSME_"+date2[:6]+".xlsx", sheet_name='Sheet1',header=0)
Rating3 = pd.read_excel (r"T:\MIB Risk Management\Credit Risk Analytics - Historical\01_Reporting\01_Monthly_Reporting\18_Rating_CBBBRSME\BRR_CBBBRSME_"+date3[:6]+".xlsx", sheet_name='Sheet1',header=0)
Rating4 = pd.read_excel (r"T:\MIB Risk Management\Credit Risk Analytics - Historical\01_Reporting\01_Monthly_Reporting\18_Rating_CBBBRSME\BRR_CBBBRSME_"+date4[:6]+".xlsx", sheet_name='Sheet1',header=0)
Rating5 = pd.read_excel (r"T:\MIB Risk Management\Credit Risk Analytics - Historical\01_Reporting\01_Monthly_Reporting\18_Rating_CBBBRSME\BRR_CBBBRSME_"+date5[:6]+".xlsx", sheet_name='Sheet1',header=0)


date_1 = 'Sep 2022'

#BB FRR
MasterBB = pd.read_excel (r"T:\MIB Risk Management\Credit Risk Analytics - Historical\02_Data_Source\02_Business_Financing\01_Portfolio\\"+date+"\\Master BB @ "+date_1+" Funded Non Funded Final to others.xlsx", sheet_name='Master Raw')


#CB BRR
#MasterCB_BRR = pd.read_excel (r'T:\MIB Risk Management\Credit Risk Analytics - Historical\02_Data_Source\12_RDMS\\'+date+'\working - SH\CRWA '+date_1+' Final (Domestic - All except Retail).xlsx', sheet_name='Context Id=162')
```

```python
BB.columns = BB.columns.str.replace("  ", "_").str.replace("\n","")
CB.columns = CB.columns.str.replace("  ", "_").str.replace("\n","")

BB1 = BB.iloc[np.where((BB.status=='03.SMA')&(BB['mmm-yy']==date[:4]+'-'+date[4:6]+'-01'))]
CB1 = CB.iloc[np.where((CB.status=='03.SMA')&(CB['mmm-yy']==date[:4]+'-'+date[4:6]+'-01'))]

BB1['Outstanding (RM Mil)'] = BB1['O/S (After RPSIA)']/(10**6)
CB1['Outstanding (RM Mil)'] = CB1['O/S (After RPSIA)']/(10**6)

BB2 = BB1.fillna(0).groupby(['GCIF','Name','ISC_NOB','ISC_Parent'])[['Outstanding (RM Mil)']].sum().reset_index().sort_values(by=['Outstanding (RM Mil)'],ascending=False)
CB2 = CB1.fillna(0).groupby(['GCIF','Name','ISC_NOB','ISC_Parent'])[['Outstanding (RM Mil)']].sum().reset_index().sort_values(by=['Outstanding (RM Mil)'],ascending=False)


Rating.columns = Rating.columns.str.strip()
Rating.columns = Rating.columns.str.upper()
Rating.columns = Rating.columns.str.replace(" ","_")

Rating1.columns = Rating1.columns.str.strip()
Rating1.columns = Rating1.columns.str.upper()
Rating1.columns = Rating1.columns.str.replace(" ","_")

Rating2.columns = Rating2.columns.str.strip()
Rating2.columns = Rating2.columns.str.upper()
Rating2.columns = Rating2.columns.str.replace(" ","_")

Rating3.columns = Rating3.columns.str.strip()
Rating3.columns = Rating3.columns.str.upper()
Rating3.columns = Rating3.columns.str.replace(" ","_")

Rating4.columns = Rating4.columns.str.strip()
Rating4.columns = Rating4.columns.str.upper()
Rating4.columns = Rating4.columns.str.replace(" ","_")

Rating5.columns = Rating5.columns.str.strip()
Rating5.columns = Rating5.columns.str.upper()
Rating5.columns = Rating5.columns.str.replace(" ","_")




Rating_ = Rating[['M_CUS_NO','FINAL_RATING']].rename(columns={'M_CUS_NO':'GCIF','FINAL_RATING':'FINAL_RATING'+date[:6]})
Rating_['GCIF'] = Rating_['GCIF'].astype('Int64').astype('str')

Rating1_ = Rating1[['M_CUS_NO','FINAL_RATING']].rename(columns={'M_CUS_NO':'GCIF','FINAL_RATING':'FINAL_RATING'+date1[:6]})
Rating1_['GCIF'] = Rating1_['GCIF'].astype('Int64').astype('str')

Rating2_ = Rating2[['M_CUS_NO','FINAL_RATING']].rename(columns={'M_CUS_NO':'GCIF','FINAL_RATING':'FINAL_RATING'+date2[:6]})
Rating2_['GCIF'] = Rating2_['GCIF'].astype('Int64').astype('str')

Rating3_ = Rating3[['M_CUS_NO','FINAL_RATING']].rename(columns={'M_CUS_NO':'GCIF','FINAL_RATING':'FINAL_RATING'+date3[:6]})
Rating3_['GCIF'] = Rating3_['GCIF'].astype('Int64').astype('str')

Rating4_ = Rating4[['M_CUS_NO','FINAL_RATING']].rename(columns={'M_CUS_NO':'GCIF','FINAL_RATING':'FINAL_RATING'+date4[:6]})
Rating4_['GCIF'] = Rating4_['GCIF'].astype('Int64').astype('str')

Rating5_ = Rating5[['M_CUS_NO','FINAL_RATING']].rename(columns={'M_CUS_NO':'GCIF','FINAL_RATING':'FINAL_RATING'+date5[:6]})
Rating5_['GCIF'] = Rating5_['GCIF'].astype('Int64').astype('str')


BB2.GCIF = BB2.GCIF.astype('Int64').astype('str')

BB3 = BB2.merge(Rating5_.drop_duplicates(), on='GCIF', how='left')\
.merge(Rating4_.drop_duplicates(), on='GCIF', how='left')\
.merge(Rating3_.drop_duplicates(), on='GCIF', how='left')\
.merge(Rating2_.drop_duplicates(), on='GCIF', how='left')\
.merge(Rating1_.drop_duplicates(), on='GCIF', how='left')\
.merge(Rating_.drop_duplicates(), on='GCIF', how='left')



CB2.GCIF = CB2.GCIF.astype('Int64').astype('str')

CB3 = CB2.merge(Rating5_.drop_duplicates(), on='GCIF', how='left')\
.merge(Rating4_.drop_duplicates(), on='GCIF', how='left')\
.merge(Rating3_.drop_duplicates(), on='GCIF', how='left')\
.merge(Rating2_.drop_duplicates(), on='GCIF', how='left')\
.merge(Rating1_.drop_duplicates(), on='GCIF', how='left')\
.merge(Rating_.drop_duplicates(), on='GCIF', how='left')


#FRR
MasterBB1 = MasterBB[['GCIF','FRR']]

MasterBB1['GCIF'] = MasterBB1['GCIF'].astype('Int64').astype('str')

BB4 = BB3.merge(MasterBB1.drop_duplicates(), on='GCIF', how='left')


MasterCB_FRR1 = MasterCB_FRR[['GCIF_NUM','FIN_FRR']].rename(columns={'GCIF_NUM':'GCIF'})

MasterCB_FRR1['GCIF'] = MasterCB_FRR1['GCIF'].astype('Int64').astype('str')

CB4 = CB3.merge(MasterCB_FRR1.drop_duplicates(), on='GCIF', how='left')


BB_SMA_Action_Plan1 = SMA_Action_Plan[['GCIF','Remarks(Action Plan)']]

BB_SMA_Action_Plan1['GCIF'] = BB_SMA_Action_Plan1['GCIF'].astype('Int64').astype('str')

BB5 = BB4.merge(BB_SMA_Action_Plan1, on='GCIF', how='left')


CB_SMA_Action_Plan1 = CB_SMA_Action_Plan[['GCIF','REASON FOR SMA CLASSIFICATION','ACTION PLAN']]

CB_SMA_Action_Plan1['GCIF'] = CB_SMA_Action_Plan1['GCIF'].astype('Int64').astype('str')

CB5 = CB4.merge(CB_SMA_Action_Plan1, on='GCIF', how='left')
```

```python
writer2 = pd.ExcelWriter(r'T:\MIB Risk Management\Credit Risk Analytics - Historical\01_Reporting\01_Monthly_Reporting\10_Credit_Risk_Profile\\'+date[:4]+'\\'+date[:6]+'\Top_10_SMA\working - SH\Top10 SMA working '+date[:6]+'(python).xlsx',engine='xlsxwriter')

CB5.to_excel(writer2, sheet_name='CB',index = False)
BB5.to_excel(writer2, sheet_name='BB', index = False)

writer2.save()
```
