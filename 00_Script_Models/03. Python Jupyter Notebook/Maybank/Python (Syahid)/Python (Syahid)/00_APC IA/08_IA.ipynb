```python
TOTAL_IMPAIRED_FINAL1 = TOTAL_IMPAIRED_FINAL.rename(columns={'V_ORIGINAL_ACCOUNT_NUMBER':'M_ACCOUNT_NO'})

TOTAL_IMPAIRED_FINAL1.loc[TOTAL_IMPAIRED_FINAL1.PROD_LV1=='On Balance Sheet','IMPAIRED_FLAG']='Y'

TOTAL_IMPAIRED_FINAL1.M_ACCOUNT_NO = TOTAL_IMPAIRED_FINAL1.M_ACCOUNT_NO.astype(str)
TOTAL_IMPAIRED_FINAL1.M_ACCOUNT_NO = TOTAL_IMPAIRED_FINAL1.M_ACCOUNT_NO.str.upper()
TOTAL_IMPAIRED_FINAL1.M_ACCOUNT_NO = TOTAL_IMPAIRED_FINAL1.M_ACCOUNT_NO.str.strip()

A001.M_ACCOUNT_NO = A001.M_ACCOUNT_NO.astype(str)
A001.M_ACCOUNT_NO = A001.M_ACCOUNT_NO.str.upper()
A001.M_ACCOUNT_NO = A001.M_ACCOUNT_NO.str.strip()

A012 = A001.merge(TOTAL_IMPAIRED_FINAL1[['M_ACCOUNT_NO','IMPAIRED_FLAG','F_NCIF_CUST','F_NCIF_ACCT','OUTSTANDING_AFTER_EIR_UWI']],on='M_ACCOUNT_NO',how='left')


A012.fillna(0,inplace=True)

A022 = A012.groupby([
'DATA_SOURCE'\
,'M_ACCOUNT_NO'\
,'M_CUS_NO'\
,'LEVEL_6_ORI'\
,'LEVEL_6'\
,'M_SUB_MARKET_SEGMENT_CD'\
,'M_PRODUCT_HIERARCHY_CD'\
,'M_SOURCE_PROD_TYPE_CD'\
,'SOURCE_PROD_TYPE_DSC'\
,'M_MTHS_IN_ARREARS'\
,'APPROACH'\
,'SECTOR_CATEGORY'\
,'PD_SEGMENT'\
,'RSME_INCLUSION_INDICATOR'\
,'RS_INCLUSION_INDICATOR'\
,'REPORT_INDICATOR'\
,'F_RPSIA'\
,'RPSIA_THIRD_PARTY'\
,'IMPAIRED_FLAG'\
,'F_NCIF_CUST'\
,'F_NCIF_ACCT'\
,'SECURITY_TYPE_CD'\
,'M_FINANCING_CONCEPT'])[['OUTSTANDING_AFTER_EIR_UWI'\
,'GROSS_LOAN_RPSIA'\
,'M_BNM_BALANCE2'\
,'EAD_PROPORTIONAL_MYR'\
,'EXPECTED_LOSS_MYR'\
,'RWA_AFTER_RISK_MIT_MYR']].sum().reset_index()


A001 = A022.fillna(0)

def PRODUCT_GROUP_APC(DATA_SOURCE,LEVEL_6,M_Sub_Market_Segment_Cd,M_Product_Hierarchy_Cd):
    #NON-RETAIL FINANCING
    if( ((DATA_SOURCE!='07. HP')&(DATA_SOURCE!='08. CC'))&\
       ((DATA_SOURCE == '01. LOAN')|(DATA_SOURCE == '02. OD')|(DATA_SOURCE == '03. STRC')|(DATA_SOURCE == '04. INV')|\
        (DATA_SOURCE == '05. TRADE')|(DATA_SOURCE == '06. FCL'))&((LEVEL_6!='Salary')&(LEVEL_6!='SMF'))):
        if((M_Sub_Market_Segment_Cd == 5)|(M_Sub_Market_Segment_Cd == 6)|(M_Sub_Market_Segment_Cd == 7)|\
           (M_Sub_Market_Segment_Cd == 8))|((M_Product_Hierarchy_Cd=='HZK01')|(M_Product_Hierarchy_Cd=='HZK02')|\
                                        (M_Product_Hierarchy_Cd=='HZK03')|(M_Product_Hierarchy_Cd=='HZK04')):
            return '01. Corporate Banking'
        elif(M_Sub_Market_Segment_Cd == 2):
            return '02. Business Banking'
        elif(M_Sub_Market_Segment_Cd == 1):
            return '03. RSME Banking'
    #RETAIL FINANCING
    #MORTGAGE FINANCING
    if(((DATA_SOURCE == '01. LOAN')|(DATA_SOURCE == '02. OD')|(DATA_SOURCE == '06. FCL'))&((M_Sub_Market_Segment_Cd == 0)|(M_Sub_Market_Segment_Cd == 3)|\
                                                                                           (M_Sub_Market_Segment_Cd == 4))):
        if(LEVEL_6 == 'Housing'):
            return '04. Home Financing'
        elif(LEVEL_6 == 'Shophouse'):
            return '05. Shophouse'
        elif(LEVEL_6 == 'Other Mortgage Loan'):
            return '06. Other Mortgage'
    #HIRE PURCHASE
    if(DATA_SOURCE == '07. HP'):
        return '07. Hire Purchase'
    #PRODUCT EXCLUDE FROM PERSONAL FINANCING ONLY WO CARDS
    if((DATA_SOURCE == '01. LOAN')&((M_Product_Hierarchy_Cd == '711')|(M_Product_Hierarchy_Cd == '714'))):
        return '08. Credit Cards'
    if(DATA_SOURCE == '08. CC'):
        return '08. Credit Cards'
    #UNIT TRUST
    if((DATA_SOURCE == '01. LOAN')&((M_Sub_Market_Segment_Cd == 0)|(M_Sub_Market_Segment_Cd == 3)|(M_Sub_Market_Segment_Cd == 4))):
        if((LEVEL_6 == 'Unit Trust')|(LEVEL_6 == 'Unit Trust-ASB')|(LEVEL_6 == 'Unit Trust-ASB2')):
            return '09. Unit Trust'
    #SALARY
    if(LEVEL_6=='Salary'):
        return '11. Salary'
    #OTHER RETAIL FINANCING
    if((DATA_SOURCE == '01. LOAN')&((M_Sub_Market_Segment_Cd == 0)|(M_Sub_Market_Segment_Cd == 3)|(M_Sub_Market_Segment_Cd == 4))):
        if(((LEVEL_6!='Unit Trust')&(LEVEL_6!='Unit Trust-ASB')&(LEVEL_6!='Unit Trust-ASB2')&\
            (LEVEL_6!='Housing')&(LEVEL_6!='Shophouse')&(LEVEL_6!='Other Mortgage Loan')&(LEVEL_6!='Salary'))&\
           (DATA_SOURCE == '01. LOAN')&((M_Product_Hierarchy_Cd!='711')&(M_Product_Hierarchy_Cd!='714')&\
                                        (M_Product_Hierarchy_Cd!='HZK01')&(M_Product_Hierarchy_Cd!='HZK02')&\
                                        (M_Product_Hierarchy_Cd!='HZK03')&(M_Product_Hierarchy_Cd!='HZK04'))):
            return '12. Other Retail Financing'
    #CASHLINE
    if((DATA_SOURCE=='02. OD')&((M_Sub_Market_Segment_Cd == 0)|(M_Sub_Market_Segment_Cd == 3)|(M_Sub_Market_Segment_Cd == 4))&\
       ((LEVEL_6!='SMF')&(LEVEL_6!='Housing')&(LEVEL_6!='Shophouse')&(LEVEL_6!='Other Mortgage Loan'))):
        return '13. Cashline'
    #SMF
    if(LEVEL_6=='SMF'):
        return '14. SMF'
A001['PRODUCT_GROUP_APC'] = A001.apply(lambda x: PRODUCT_GROUP_APC(x['DATA_SOURCE'], x['LEVEL_6'],x['M_SUB_MARKET_SEGMENT_CD'],x['M_PRODUCT_HIERARCHY_CD']), axis=1)

#==============================================================================

def ASSET_CLASS(Approach,Sector_Category,Rsme_Inclusion_Indicator,Rs_Inclusion_Indicator,Report_Indicator):
    if(Approach == 2)&((Sector_Category == 1)|(Sector_Category == 2)):
        return 'SOVEREIGN'
    elif(Sector_Category==4): 
        return 'PSE' #APPROACH = 2 AND -changes on August 2018 report only
    elif(Approach==3)&((Sector_Category==0)|(Sector_Category==6)|(Sector_Category==7)|(Sector_Category==8)|\
                        (Sector_Category ==10)):
        return 'CORPORATE(FIRB)'
    elif(Approach==2)&((Sector_Category==0)|(Sector_Category==6)|(Sector_Category==7)|(Sector_Category==8)|\
                        (Sector_Category ==10))&(Report_Indicator==90): 
        return 'CORPORATE(FIRB)'
    elif(Approach == 2)&((Sector_Category==0)|(Sector_Category==6)|(Sector_Category==7)|(Sector_Category==8)|\
                        (Sector_Category ==10)):
        return 'CORPORATE(SA)'
    elif(Approach==6)&(Rs_Inclusion_Indicator!=0)&(Rs_Inclusion_Indicator!='0'):
        return 'RETAIL (AIRB)'
    elif(Approach==6)&(Rsme_Inclusion_Indicator!='0')&(Rsme_Inclusion_Indicator!=0):
        return 'RSME (AIRB)'
    elif(Approach==2)&(Sector_Category==9):
        return 'RETAIL (SA)'
    elif(Approach==3)&(Sector_Category==5):
        return 'BANK(FIRB)'
    elif(Approach==2)&(Sector_Category==5):
        return 'BANK(SA)' #added on 14 Jan 2020
A001['ASSET_CLASS'] = A001.apply(lambda x: ASSET_CLASS(x['APPROACH'],x['SECTOR_CATEGORY'],x['RSME_INCLUSION_INDICATOR'],x['RS_INCLUSION_INDICATOR'],x['REPORT_INDICATOR']), axis=1)


A001.loc[A001.LEVEL_6 == 'Unit Trust-ASB', 'P2_RWA_MIL'] = (A001.EAD_PROPORTIONAL_MYR - A001.RWA_AFTER_RISK_MIT_MYR)/1000000
A001.loc[A001.LEVEL_6 != 'Unit Trust-ASB', 'P2_RWA_MIL'] = 0

A001['M_BNM_BALANCE_MIL'] = A001['M_BNM_BALANCE2']/1000000
A001['EAD_PROPORTIONAL_MYR_MIL'] = A001['EAD_PROPORTIONAL_MYR']/1000000
A001['EXPECTED_LOSS_MYR_MIL'] = A001['EXPECTED_LOSS_MYR']/1000000
A001['RWA_AFTER_RISK_MIT_MYR_MIL'] = A001['RWA_AFTER_RISK_MIT_MYR']/1000000
A001['OUTSTANDING_AFTER_EIR_UWI_MIL'] = A001['OUTSTANDING_AFTER_EIR_UWI']/1000000
A001['GROSS_LOAN_RPSIA_MIL'] =  A001['GROSS_LOAN_RPSIA']/1000000
```
