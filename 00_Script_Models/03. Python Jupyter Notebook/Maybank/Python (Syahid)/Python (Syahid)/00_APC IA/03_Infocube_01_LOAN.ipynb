```python
date = 20220930
nYear = 365

#LDAY = 30 #June

import numpy as np
import pandas as pd
import pandasql
import sqlite3

pd.set_option("display.max_columns", None) 
pd.set_option("display.max_colwidth", 1000) #huruf dlm column
pd.set_option("display.max_rows", 100)
pd.set_option("display.precision", 2) #2 titik perpuluhan


#Teradata cust
location_cust = r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\02_Teradata\01_Credit_Risk_Reporting\01_Cust\\"+str(date)+"\\S001_004_ACRM_CUST"+str(date)+".txt"
cust = pd.read_csv(location_cust, sep = "|", header = 0, error_bad_lines = False)
cust.columns = cust.columns.str.replace(" ", "_")
cust.M_SUB_SUB_MARKET_SEGMENT.replace({'    ': -9999}, inplace=True)

#Customer Industry
location_cust_ind = r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\02_Teradata\01_Credit_Risk_Reporting\02_Cust_Ind\\"+str(date)+"\\S001_028_ACRM_CUST_IND"+str(date)+".txt"
cust_ind = pd.read_csv(location_cust_ind, sep = "|", header = 0)
cust_ind.columns = cust_ind.columns.str.replace(" ", "_")

#Organization
location_org = r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\02_Teradata\01_Credit_Risk_Reporting\04_Organization\\"+str(date)+"\\T0120_ORGANIZATION"+str(date)+".txt"
org = pd.read_csv(location_org, sep = "|", header = 0)
org.columns = org.columns.str.replace(" ", "_")

#RDMS
location_RDMS = r'T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\01_Datawarehouse\01_Credit_Risk_Reporting\04_MIB_Rdms\\'+str(date)+'\\RDMS_'+str(date)+'.txt'
RDMS = pd.read_csv(location_RDMS, sep = ",", header = 0, low_memory = False)
RDMS.columns = RDMS.columns.str.replace(" ", "_")


#-----------------------------------------------------USER_DEFINED---------------------------------------------------

Branch_Dsc = pd.read_csv(r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\01_Datawarehouse\02_User_Defined\01_Branch_Dsc\Branch_Dsc.txt", sep = "\t", header = 0, low_memory = False)
Branch_Dsc.columns = Branch_Dsc.columns.str.replace(" ", "_")

Product_Hierarchy = pd.read_csv (r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\01_Datawarehouse\02_User_Defined\06_Product_Hierarchy\Product_Hierarchy.txt", sep = "\t", header = 0, low_memory = False)
Product_Hierarchy.columns = Product_Hierarchy.columns.str.replace(" ", "_")

Econ_Sector_Dsc = pd.read_csv(r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\01_Datawarehouse\02_User_Defined\02_Econ_Sector_Dsc\Econ_Sector_Dsc.txt", sep = "\t", header = 0, low_memory = False)
Econ_Sector_Dsc.columns = Econ_Sector_Dsc.columns.str.replace(" ", "_")

Income_Grp_Dsc = pd.read_csv(r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\01_Datawarehouse\02_User_Defined\03_Income_Grp_Dsc\Income_Grp_Dsc.txt", sep = "\t", header = 0, low_memory = False)
Income_Grp_Dsc.columns = Income_Grp_Dsc.columns.str.replace(" ", "_")

Interest_Rate_Cd_Maps = pd.read_csv(r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\01_Datawarehouse\02_User_Defined\08_Interest_Rate_Cd_Maps\Interest_Rate_Cd_Maps.txt", sep = "\t", header = 0, low_memory = False)
Interest_Rate_Cd_Maps.columns = Interest_Rate_Cd_Maps.columns.str.replace(" ", "_")

Occ_Sector_Dsc = pd.read_csv(r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\01_Datawarehouse\02_User_Defined\05_Occ_Sector_Dsc\new version\Occ_Sector_Dsc.txt", sep = "\t", header = 0, low_memory = False)
Occ_Sector_Dsc.columns = Occ_Sector_Dsc.columns.str.replace(" ", "_")

Sub_Mkt_Dsc = pd.read_csv(r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\01_Datawarehouse\02_User_Defined\07_Sub_Mkt_Dsc\Sub_Mkt_Dsc.txt", sep = "\t", header = 0, low_memory = False)
Sub_Mkt_Dsc.columns = Sub_Mkt_Dsc.columns.str.replace(" ", "_")

R028_NATIONALITY_CODE = pd.read_csv(r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\02_Teradata\02_User_Defined\01_Nationality_Code\R028_NATIONALITY_CODE.txt", sep = "\t", header = 0, low_memory = False)
R028_NATIONALITY_CODE.columns = R028_NATIONALITY_CODE.columns.str.replace(" ", "_")

#------------------------------------------------------info01---------------------------------------------------------

#Loans
location_loan = r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\01_Datawarehouse\01_Credit_Risk_Reporting\01_MIB_Loans\\"+str(date)+"\\MIB_Loans"+str(date)+".txt"
loan = pd.read_csv(location_loan, sep = "\t", header = 0, low_memory = False)
loan.columns = loan.columns.str.replace(" ", "_")

loan.fillna(0, inplace =True)

loan['M_Account_No'] = loan['M_Account_No'].astype(str)
loan['M_Account_No'] = loan['M_Account_No'].str[:20]

loan['M_Cus_No'] = loan['M_Cus_No'].astype(str)
loan['M_Cus_No'] = loan['M_Cus_No'].str[:10]

loan['M_Source_Prod_Type_Cd'] = loan['M_Source_Prod_Type_Cd'].astype(str)
loan['M_Source_Prod_Type_Cd'] = loan['M_Source_Prod_Type_Cd'].str[:15]

loan['M_Approved_Limit_SUM'] = loan['M_Approved_Limit_SUM'].astype(float)

loan['M_Bnm_Balance_SUM'] = loan['M_Bnm_Balance_SUM'].astype(float)

loan['M_Product_Hierarchy_Cd'] = loan['M_Product_Hierarchy_Cd'].astype(str)
loan['M_Product_Hierarchy_Cd'] = loan['M_Product_Hierarchy_Cd'].str[:15]

loan['M_Economic_Sector_Cd'] = loan['M_Economic_Sector_Cd'].astype(str)
loan['M_Economic_Sector_Cd'] = loan['M_Economic_Sector_Cd'].str[:8]

loan['Amrt_Term'] = loan['Amrt_Term'].astype(float)

loan['M_Mths_Installment_In_Arrears'] = loan['M_Mths_Installment_In_Arrears'].astype(float)

loan['M_Reschedule_Flg'] = loan['M_Reschedule_Flg'].astype(str)
loan['M_Reschedule_Flg'] = loan['M_Reschedule_Flg'].str[:1]

loan['M_Restructure_Flg'] = loan['M_Restructure_Flg'].astype(str)
loan['M_Restructure_Flg'] = loan['M_Restructure_Flg'].str[:1]

loan['M_Interest_Income_SUM'] = loan['M_Interest_Income_SUM'].astype(float)

loan['M_Non_Interest_Income_SUM'] = loan['M_Non_Interest_Income_SUM'].astype(float)

loan['Cur_Gross_Rate_SUM'] = loan['Cur_Gross_Rate_SUM'].astype(float)

loan['M_Penalty_Charges_SUM'] = loan['M_Penalty_Charges_SUM'].astype(float)

loan['M_Misc_Charges_SUM'] = loan['M_Misc_Charges_SUM'].astype(float)

loan['M_Funds_Charge_Myr_SUM'] = loan['M_Funds_Charge_Myr_SUM'].astype(float)
  
loan['M_Interest_Income_Myr_SUM'] = loan['M_Interest_Income_Myr_SUM'].astype(float)

loan['M_Net_Int_Income_Myr_SUM'] = loan['M_Net_Int_Income_Myr_SUM'].astype(float)

loan['M_New_Npl_Flg'] = loan['M_New_Npl_Flg'].astype(str)
loan['M_New_Npl_Flg'] = loan['M_New_Npl_Flg'].str[:1]

loan['M_Entity_Cd'] = loan['M_Entity_Cd'].astype(str)
loan['M_Entity_Cd'] = loan['M_Entity_Cd'].str[:3]

loan['M_Acct_Status_Cd'] = loan['M_Acct_Status_Cd'].astype(float)

loan['M_Npl_Flg'] = loan['M_Npl_Flg'].astype(str)
loan['M_Npl_Flg'] = loan['M_Npl_Flg'].str[:1]

loan['Avg_Book_Bal_SUM'] = loan['Avg_Book_Bal_SUM'].astype(float)

loan['Adjustable_Type_Cd'] = loan['Adjustable_Type_Cd'].astype(float)

loan['M_Full_Writeoff_Flg'] = loan['M_Full_Writeoff_Flg'].astype(str)
loan['M_Full_Writeoff_Flg'] = loan['M_Full_Writeoff_Flg'].str[:1]

loan['M_Interest_Rate_Type'] = loan['M_Interest_Rate_Type'].astype(str)
loan['M_Interest_Rate_Type'] = loan['M_Interest_Rate_Type'].str[:1]

loan['M_Aa_Number'] = loan['M_Aa_Number'].astype(str)
loan['M_Aa_Number'] = loan['M_Aa_Number'].str[:30]

loan['Amrt_Term_Mult'] = loan['Amrt_Term_Mult'].astype(str)
loan['Amrt_Term_Mult'] = loan['Amrt_Term_Mult'].str[:1]

loan['M_Financing_Concept'] = loan['M_Financing_Concept'].astype(str)
loan['M_Financing_Concept'] = loan['M_Financing_Concept'].str[:3]


RDMS_1 = RDMS[['Account_No','Ead_Proportional_Myr','Expected_Loss_Myr','Rwa_After_Risk_Mit_Myr','Approach','Sector_Category','Rsme_Inclusion_Indicator','Rs_Inclusion_Indicator','Report_Indicator']]

RDMS_1.Account_No = RDMS_1.Account_No.astype(str)
RDMS_1.Account_No = RDMS_1.Account_No.str.upper()
RDMS_1.Account_No = RDMS_1.Account_No.str.strip()

loan.M_Account_No = loan.M_Account_No.astype(str)
loan.M_Account_No = loan.M_Account_No.str.upper()
loan.M_Account_No = loan.M_Account_No.str.strip()


LN002 = loan.merge(RDMS_1[['Account_No','Ead_Proportional_Myr','Expected_Loss_Myr','Rwa_After_Risk_Mit_Myr','Approach','Sector_Category','Rsme_Inclusion_Indicator','Rs_Inclusion_Indicator','Report_Indicator']].rename(columns={'Account_No':'M_Account_No'}),on='M_Account_No', how='left')

LN002 = LN002.drop_duplicates()

#----------------------------------------LN003---------------------------------------------

RDMS_2 = RDMS[['Account_No','Rsme_Rating_Grade','Rating','Pd_Segment_Id']]

LN002.M_Account_No = LN002.M_Account_No.astype(str)
LN002.M_Account_No = LN002.M_Account_No.str.upper()
LN002.M_Account_No = LN002.M_Account_No.str.strip()

RDMS_2.Account_No = RDMS_2.Account_No.astype(str)
RDMS_2.Account_No = RDMS_2.Account_No.str.upper()
RDMS_2.Account_No = RDMS_2.Account_No.str.strip()

LN003 = LN002.merge(RDMS_2.rename(columns={'Account_No': 'M_Account_No'}), on = 'M_Account_No', how = 'left')

LN003.loc[LN003.M_Sub_Market_Segment_Cd.isin([1]), 'PD_SEGMENT'] = LN003['Rsme_Rating_Grade']
LN003.loc[LN003.M_Sub_Market_Segment_Cd.isin([2,5,6,7,8]), 'PD_SEGMENT'] = LN003['Rating']
LN003.loc[~(LN003.M_Sub_Market_Segment_Cd.isin([1,2,5,6,7,8])), 'PD_SEGMENT'] = LN003['Pd_Segment_Id']

LN003_1 = LN003.drop(columns=['Rsme_Rating_Grade','Rating','Pd_Segment_Id'])

LN003_1.sort_values(by=['M_Product_Hierarchy_Cd'],ascending=True, inplace=True) 


#----------------------------------------#LEVEL 6#----------------------------------------

PRD = Product_Hierarchy[['Product_Hierarchy_Cd','Level_6','Level_5']]
PRD = PRD.sort_values(by=['Product_Hierarchy_Cd'], ascending=True)

PRD.Product_Hierarchy_Cd = PRD.Product_Hierarchy_Cd.astype(str)
PRD.Product_Hierarchy_Cd = PRD.Product_Hierarchy_Cd.str.upper()
PRD.Product_Hierarchy_Cd = PRD.Product_Hierarchy_Cd.str.strip()

LN003_1.M_Product_Hierarchy_Cd = LN003_1.M_Product_Hierarchy_Cd.astype(str)
LN003_1.M_Product_Hierarchy_Cd = LN003_1.M_Product_Hierarchy_Cd.str.upper()
LN003_1.M_Product_Hierarchy_Cd = LN003_1.M_Product_Hierarchy_Cd.str.strip()

LN004 = LN003_1.merge(PRD.rename(columns={'Product_Hierarchy_Cd': 'M_Product_Hierarchy_Cd'}), on = 'M_Product_Hierarchy_Cd', how = 'left')

LN004.sort_values(by=["M_Economic_Sector_Cd"],ascending=True, inplace=True) 

#----------------------------------------#ECONOMIC SECTOR#----------------------------------------

ECON = Econ_Sector_Dsc[["Econ_Sector_Cd", "Econ_Sector"]]

ECON.Econ_Sector_Cd = ECON.Econ_Sector_Cd.astype(str)
ECON.Econ_Sector_Cd = ECON.Econ_Sector_Cd.str.upper()
ECON.Econ_Sector_Cd = ECON.Econ_Sector_Cd.str.strip()

ECON = ECON.drop_duplicates() 

LN004.M_Economic_Sector_Cd = LN004.M_Economic_Sector_Cd.astype(str)
LN004.M_Economic_Sector_Cd = LN004.M_Economic_Sector_Cd.str.upper()
LN004.M_Economic_Sector_Cd = LN004.M_Economic_Sector_Cd.str.strip()

LN005 = LN004.merge(ECON.rename(columns={'Econ_Sector_Cd': 'M_Economic_Sector_Cd'}), on = 'M_Economic_Sector_Cd', how = 'left')

LN005.sort_values(by=["Branch_Code"],ascending=True, inplace=True) 

#----------------------------------------#STATE#----------------------------------------

BRANCH = Branch_Dsc[["Region_Desc","State_Desc","Branch_Desc","Branch_Cd"]]

BRANCH.Branch_Cd = BRANCH.Branch_Cd.astype(str)
BRANCH.Branch_Cd = BRANCH.Branch_Cd.str.upper()
BRANCH.Branch_Cd = BRANCH.Branch_Cd.str.strip()

LN005.Branch_Code = LN005.Branch_Code.astype(str)
LN005.Branch_Code = LN005.Branch_Code.str.upper()
LN005.Branch_Code = LN005.Branch_Code.str.strip()

LN006 = LN005.merge(BRANCH.rename(columns={"Branch_Cd": "Branch_Code"}), on = 'Branch_Code', how = 'left')

LN006.sort_values(by=["M_Cus_No"], ascending=True, inplace=True) 

#----------------------------------------#CUSTOMER SEGMENT#----------------------------------------

cust_1 = cust[["CUS_NO" ,"CUSTOMER_SEGMENT"]]

cust_1.CUS_NO = cust_1.CUS_NO.astype(str)
cust_1.CUS_NO = cust_1.CUS_NO.str.upper()
cust_1.CUS_NO = cust_1.CUS_NO.str.strip()

LN006.M_Cus_No = LN006.M_Cus_No.astype(str)
LN006.M_Cus_No = LN006.M_Cus_No.str.upper()
LN006.M_Cus_No = LN006.M_Cus_No.str.strip()

LN007 = LN006.merge(cust_1.rename(columns={"CUS_NO": "M_Cus_No"}), on = 'M_Cus_No', how = 'left')

LN007.CUSTOMER_SEGMENT.replace({'   ': 0}, inplace=True)
LN007.sort_values(by=["M_Cus_No"], ascending=True, inplace=True) 

#-----------------#RACE CD & SEX & NATIONALITY CD & BIRTH DATE & OCCUPATION CODE & INCOME GROUP CD & MARRIED

cust_ind_1 = cust_ind.drop(columns=["AS_OF_DATE"])

cust_ind_1.CUS_NO = cust_ind_1.CUS_NO.astype(str)
cust_ind_1.CUS_NO = cust_ind_1.CUS_NO.str.upper()
cust_ind_1.CUS_NO = cust_ind_1.CUS_NO.str.strip()

LN007.M_Cus_No = LN007.M_Cus_No.astype(str)
LN007.M_Cus_No = LN007.M_Cus_No.str.upper()
LN007.M_Cus_No = LN007.M_Cus_No.str.strip()

LN008 = LN007.merge(cust_ind_1.rename(columns={"CUS_NO": "M_Cus_No"}), on = 'M_Cus_No', how = 'left')

LN008.sort_values(by=['M_NATIONALITY_CD'], ascending=True, inplace=True)

#----------------------------------------#NATIONALITY#----------------------------------------

NATION = R028_NATIONALITY_CODE[['MBB_Nationality_Cd' , 'MBB_Nationality_Desc']]

NATION.MBB_Nationality_Cd = NATION.MBB_Nationality_Cd.astype(str)
NATION.MBB_Nationality_Cd = NATION.MBB_Nationality_Cd.str.upper()
NATION.MBB_Nationality_Cd = NATION.MBB_Nationality_Cd.str.strip()

LN008.M_NATIONALITY_CD = LN008.M_NATIONALITY_CD.astype(str)
LN008.M_NATIONALITY_CD = LN008.M_NATIONALITY_CD.str.upper()
LN008.M_NATIONALITY_CD = LN008.M_NATIONALITY_CD.str.strip()

LN009 = LN008.merge(NATION.rename(columns={'MBB_Nationality_Cd': 'M_NATIONALITY_CD'}), on = 'M_NATIONALITY_CD', how = 'left')

#----------------------------------------#OCCUPATION SECTOR#----------------------------------------

OCC = Occ_Sector_Dsc[["Occ_Sector","Occ_Sector_Cd"]]

OCC.Occ_Sector_Cd = OCC.Occ_Sector_Cd.astype(str)
OCC.Occ_Sector_Cd = OCC.Occ_Sector_Cd.str.upper()
OCC.Occ_Sector_Cd = OCC.Occ_Sector_Cd.str.strip()

LN009.M_OCC_SECTOR_CD = LN009.M_OCC_SECTOR_CD.astype(str)
LN009.M_OCC_SECTOR_CD = LN009.M_OCC_SECTOR_CD.str.upper()
LN009.M_OCC_SECTOR_CD = LN009.M_OCC_SECTOR_CD.str.strip()

LN010 = LN009.merge(OCC.rename(columns={"Occ_Sector_Cd": "M_OCC_SECTOR_CD"}), on = 'M_OCC_SECTOR_CD', how = 'left')

LN010.M_OCC_SECTOR_CD.replace({'        ': 0}, inplace=True)
LN010.sort_values(by=['M_INCOME_GRP_CD'], ascending=True, inplace=True) 

#----------------------------------------#GCIF INCOME#----------------------------------------

INCOM = Income_Grp_Dsc[['Income_Grp', 'Income_Grp_Cd']]

INCOM.Income_Grp_Cd = INCOM.Income_Grp_Cd.astype(str)
INCOM.Income_Grp_Cd = INCOM.Income_Grp_Cd.str.upper()
INCOM.Income_Grp_Cd = INCOM.Income_Grp_Cd.str.strip()

LN010.M_INCOME_GRP_CD = LN010.M_INCOME_GRP_CD.astype(str)
LN010.M_INCOME_GRP_CD = LN010.M_INCOME_GRP_CD.str.upper()
LN010.M_INCOME_GRP_CD = LN010.M_INCOME_GRP_CD.str.strip()

LN011 = LN010.merge(INCOM.rename(columns={'Income_Grp_Cd':'M_INCOME_GRP_CD'}), on = 'M_INCOME_GRP_CD', how = 'left')

LN011.sort_values(by=["M_Sub_Market_Segment_Cd"], ascending=True, inplace=True)

#----------------------------------------#M SUB MARKET SEGMENT DESCRIPTION#----------------------------------------

SUB_MKT = Sub_Mkt_Dsc[['Description', 'Code']]

SUB_MKT.Code = SUB_MKT.Code.astype(float)

SUB_MKT.rename(columns={"Code": "M_Sub_Market_Segment_Cd", "Description" : "M_Sub_Market_Segment_Desc"}, inplace=True)

LN011.M_Sub_Market_Segment_Cd = LN011.M_Sub_Market_Segment_Cd.astype(float)

LN012 = pd.merge(LN011, SUB_MKT, on = ["M_Sub_Market_Segment_Cd"], how="left")

LN012.sort_values(by=['M_Account_No'], ascending=True, inplace=True) 

#----------------------------------------#RATING(CORP) & RSME RATING GRADE#----------------------------------------

PD002 = RDMS.iloc[np.where( ~(RDMS.Rating.isnull())&(RDMS.Rating != "BKNM1") )]
PD002_1 = PD002[['Rating', 'Account_No']] 
PD002_1.sort_values(by=['Account_No'], ascending=True, inplace=True)

PD003 = RDMS.iloc[np.where(~(RDMS.Rsme_Rating_Grade.isnull()))]
PD003_1 = PD003[['Rsme_Rating_Grade','Account_No']]
PD003_1.sort_values(by=['Account_No'], ascending=True, inplace=True)

PD002_1.rename(columns={'Account_No': 'M_Account_No'}, inplace=True)
PD003_1.rename(columns={'Account_No': 'M_Account_No'}, inplace=True)

LN013 = pd.merge(LN012, PD002_1, on = ["M_Account_No"], how="left")

LN014 = pd.merge(LN013, PD003_1, on = ["M_Account_No"], how="left")

LN014.sort_values(by=['M_Source_Prod_Type_Cd'], ascending=True, inplace=True) 


#----------------------------------------#SOURCE PRODUCT TYPE DESC#----------------------------------------

PRD_2 = Product_Hierarchy[['Source_Prod_Type_Cd', 'Source_Prod_Type_Dsc']]
PRD_2 = PRD_2.drop_duplicates() 
PRD_2 = PRD_2.rename(columns={'Source_Prod_Type_Cd': 'M_Source_Prod_Type_Cd'})

PRD_2.M_Source_Prod_Type_Cd = PRD_2.M_Source_Prod_Type_Cd.astype(str)
PRD_2.M_Source_Prod_Type_Cd = PRD_2.M_Source_Prod_Type_Cd.str.upper()
PRD_2.M_Source_Prod_Type_Cd = PRD_2.M_Source_Prod_Type_Cd.str.strip()

LN014.M_Source_Prod_Type_Cd = LN014.M_Source_Prod_Type_Cd.astype(str)
LN014.M_Source_Prod_Type_Cd = LN014.M_Source_Prod_Type_Cd.str.upper()
LN014.M_Source_Prod_Type_Cd = LN014.M_Source_Prod_Type_Cd.str.strip()

LN015 = LN014.merge(PRD_2, on = ["M_Source_Prod_Type_Cd"], how="left")

LN015.sort_values(by=['M_Interest_Code'], ascending=True, inplace=True) 

#----------------------------------------#INTEREST RATE CODE DESC#----------------------------------------

INT_1 = Interest_Rate_Cd_Maps[['Code', 'Description']]
INT_1 = INT_1.drop_duplicates('Code', keep='last') 
INT_1 = INT_1.rename(columns={'Code': 'M_Interest_Code','Description':'Interest_Rate_Desc'})

INT_1.M_Interest_Code = INT_1.M_Interest_Code.astype(str)
INT_1.M_Interest_Code = INT_1.M_Interest_Code.str.upper()
INT_1.M_Interest_Code = INT_1.M_Interest_Code.str.strip()

LN015.M_Interest_Code = LN015.M_Interest_Code.astype(str)
LN015.M_Interest_Code = LN015.M_Interest_Code.str.upper()
LN015.M_Interest_Code = LN015.M_Interest_Code.str.strip()

LN016 = LN015.merge(INT_1, on = ["M_Interest_Code"], how="left")

LN016.sort_values(by=['M_Cus_No'], ascending=True, inplace=True) 

#----------------------------------------#NATURE OF BUSINESS#----------------------------------------

OG001 = org.iloc[np.where( (org.Org_Type_Cd == 'External')&(org.Record_Deleted_Flg == 0) )]
OG001.sort_values(by=['Org_Party_Id'], ascending=True, inplace=True)
OG001.rename(columns={'Org_Party_Id': 'M_Cus_No'}, inplace=True)
OG001 = OG001[['M_Cus_No','NOB_Cd']]

OG001.M_Cus_No = OG001.M_Cus_No.astype(str)
OG001.M_Cus_No = OG001.M_Cus_No.str.upper()
OG001.M_Cus_No = OG001.M_Cus_No.str.strip()

LN016.M_Cus_No = LN016.M_Cus_No.astype(str)
LN016.M_Cus_No = LN016.M_Cus_No.str.upper()
LN016.M_Cus_No = LN016.M_Cus_No.str.strip()

LN017 = LN016.merge(OG001, on = ["M_Cus_No"], how="left")

#============================================================================

def Race(M_RACE_CD):
    if M_RACE_CD==1:
        return 'Malay'
    elif M_RACE_CD==2:
        return 'Native'
    elif M_RACE_CD==3:
        return 'Chinese'
    elif M_RACE_CD==4:
        return 'Indian'
    elif M_RACE_CD==99:
        return 'Others'
    else: 
        return 'Not Populated'
LN017['RACE'] = LN017['M_RACE_CD'].apply(Race)


def Gender(SEX):
    if SEX==1:
        return 'Female'
    elif SEX==2: 
        return 'Male'
    else: 
        return 'Not Populated'
LN017['GENDER'] = LN017['SEX'].apply(Gender)


LN017.BTH_DT = LN017.BTH_DT.astype(str)
LN017['BIRTHDAYYEAR'] = LN017.BTH_DT.str[-6:]

LN017.BIRTHDAYYEAR = LN017.BIRTHDAYYEAR.astype(float)
LN017.BIRTHDAYYEAR.fillna(0, inplace=True)

def AGE(BIRTHDAYYEAR):
    if BIRTHDAYYEAR == 0:
        return 0
    else:
        return 2022 - BIRTHDAYYEAR
LN017['AGE'] = LN017['BIRTHDAYYEAR'].apply(AGE)


def AGE_RANGE(AGE):
    if AGE >= 1 and AGE < 25: 
        return 'A. <25'
    elif AGE >= 25 and AGE < 30:
        return 'B. 25-29'
    elif AGE >= 30 and AGE < 35:
        return 'C. 30-34'
    elif AGE >= 35 and AGE < 40: 
        return 'D. 35-39'
    elif AGE >= 40 and AGE < 45: 
        return 'E. 40-44'
    elif AGE >= 45 and AGE < 50: 
        return 'F. 45-49'
    elif AGE >= 50 and AGE < 55: 
        return 'G. 50-54'
    elif AGE >= 55 and AGE < 99: 
        return 'H. >=55'
    else: 
        return 'I. NOT POPULATED'
LN017['AGE_RANGE'] = LN017['AGE'].apply(AGE_RANGE)


def MARITAL_STATUS(MARRIED):
    if MARRIED=='1': 
        return 'Single'
    elif MARRIED=='2':
        return 'Married'
    else: 
        return 'Not Populated'
LN017['MARITAL_STATUS'] = LN017['MARRIED'].apply(MARITAL_STATUS)


def RELEASED_FLAG(M_Date_Final_Released):
     if M_Date_Final_Released=='01-JAN-1900':
        return 'PARTIAL RELEASED'
     else:
        return 'FULLY RELEASED'
LN017['RELEASED_FLAG'] = LN017['M_Date_Final_Released'].apply(RELEASED_FLAG)


def ADJUSTABLE_TYPE_DESC(Adjustable_Type_Cd):
    if Adjustable_Type_Cd==0:
        return 'A.Fixed Rate'
    elif Adjustable_Type_Cd==30:  
        return 'B.Administered Rate'
    elif Adjustable_Type_Cd==50:
        return 'C.Floating Rate'
    elif Adjustable_Type_Cd==250:
        return 'D.Other Adjustable'
    elif Adjustable_Type_Cd==25:
        return 'E.No Description'
LN017['ADJUSTABLE_TYPE_DESC'] = LN017['Adjustable_Type_Cd'].apply(ADJUSTABLE_TYPE_DESC)


LN017.loc[LN017.M_Interest_Code.isna(), 'Interest_Rate_Desc'] = 'BBLR'


LN017.loc[LN017.Level_5 == 'Government Aided Schemes', 'M_FUNDS_CHARGE_MYR'] = 0.015*LN017['Avg_Book_Bal_SUM']*int(str(date)[-2:])/nYear
LN017.loc[LN017.Level_5 != 'Government Aided Schemes', 'M_FUNDS_CHARGE_MYR'] = LN017['M_Funds_Charge_Myr_SUM']

LN017['NOA'] = 1


info01 = LN017.drop_duplicates('M_Account_No', keep='last')

info01.to_csv(r'T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\01_Loans\\'+str(date)+'\infocube_01_loan'+str(date)+'.txt', index = False)
```
