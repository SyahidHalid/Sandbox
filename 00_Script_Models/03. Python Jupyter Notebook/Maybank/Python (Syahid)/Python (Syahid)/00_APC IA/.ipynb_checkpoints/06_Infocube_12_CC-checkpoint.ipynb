```python
date = 20220930

import numpy as np
import pandas as pd
import pandasql
import sqlite3
pd.set_option("display.max_columns", None) 
pd.set_option("display.max_colwidth", 1000) #huruf dlm column
pd.set_option("display.max_rows", 100)
pd.set_option("display.precision", 2) #2 titik perpuluhan


location_CC = r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\01_Datawarehouse\01_Credit_Risk_Reporting\13_MIB_Credit_Cards\\"+str(date)+"\\MIB_CC"+str(date)+".txt"
CC = pd.read_csv(location_CC, sep = "\t", header = 0, error_bad_lines = False)
CC.columns = CC.columns.str.replace(" ", "_")
CC.M_Account_No = CC.M_Account_No.astype(str)
CC['M_Account_No'] = CC.M_Account_No.str[:20]


location_cust = r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\02_Teradata\01_Credit_Risk_Reporting\01_Cust\\"+str(date)+"\\S001_004_ACRM_CUST"+str(date)+".txt"
cust = pd.read_csv(location_cust, sep = "|", header = 0, error_bad_lines = False)
cust.columns = cust.columns.str.replace(" ", "_")
cust.M_SUB_SUB_MARKET_SEGMENT.replace({'    ': -9999}, inplace=True)


location_cust_ind = r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\02_Teradata\01_Credit_Risk_Reporting\02_Cust_Ind\\"+str(date)+"\\S001_028_ACRM_CUST_IND"+str(date)+".txt"
cust_ind = pd.read_csv(location_cust_ind, sep = "|", header = 0)
cust_ind.columns = cust_ind.columns.str.replace(" ", "_")


location_org = r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\02_Teradata\01_Credit_Risk_Reporting\04_Organization\\"+str(date)+"\\T0120_ORGANIZATION"+str(date)+".txt"
org = pd.read_csv(location_org, sep = "|", header = 0)
org.columns = org.columns.str.replace(" ", "_")

```

```python
location_RDMS = r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\01_Datawarehouse\01_Credit_Risk_Reporting\04_MIB_Rdms\\"+str(date)+"\RDMS_"+str(date)+".txt"
RDMS = pd.read_csv(location_RDMS, sep = ",", header = 0, low_memory = False)
RDMS.columns = RDMS.columns.str.replace(" ", "_")
RDMS.Account_No = RDMS.Account_No.astype(str)
RDMS['Account_No'] = RDMS.Account_No.str[:20]


Product_Hierarchy = pd.read_csv (r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\01_Datawarehouse\02_User_Defined\06_Product_Hierarchy\Product_Hierarchy.txt", sep = "\t", header = 0, low_memory = False)
Product_Hierarchy.columns = Product_Hierarchy.columns.str.replace(" ", "_")


Income_Grp_Dsc = pd.read_csv(r"T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\99_Data_Source\01_Datawarehouse\02_User_Defined\03_Income_Grp_Dsc\Income_Grp_Dsc.txt", sep = "\t", header = 0, low_memory = False)
Income_Grp_Dsc.columns = Income_Grp_Dsc.columns.str.replace(" ", "_")
```

```python
CC1 = CC.drop_duplicates()

CC1['M_Bnm_Balance2'] = CC1.M_Bnm_Balance

CC1_2 = CC1.drop('M_Bnm_Balance2',axis=1)

CC1_1 = CC1.groupby(['M_Account_No'])[['M_Bnm_Balance2']].sum().reset_index()

CC1_3 = CC1_2.merge(CC1_1,on='M_Account_No',how='left')
```

```python
CC02b = CC1_3.fillna(0)

CC02b.loc[(CC02b.M_Card_Block_Cd.isin(['D','L','X','Z']))&(CC02b.M_Bnm_Balance2==0),"close_tag"] = 1
CC02b.close_tag.fillna(0, inplace=True)

#CC02.loc[~((CC02.M_Card_Block_Cd.isin(['D','L','X','Z']))&(CC02.M_Bnm_Balance2==0)),"close_tag"] = 0

CC02a = CC02b.iloc[np.where((CC02b.Record_Count==1)&(CC02b.close_tag==0))]

CC02 = CC02a

RDMS_1 = RDMS[['Account_No','Ead_Proportional_Myr','Expected_Loss_Myr','Rwa_After_Risk_Mit_Myr','Pd_Segment_Id','Rating','Rsme_Rating_Grade','Approach','Sector_Category','Rsme_Inclusion_Indicator','Rs_Inclusion_Indicator','Report_Indicator']]

CC03 = CC02.merge(RDMS_1.rename(columns={'Account_No':'M_Account_No','Pd_Segment_Id':'PD_SEGMENT'}),on = 'M_Account_No', how = 'left')


CC03.sort_values(by=['M_Cus_No'], ascending=True, inplace=True)

#RACE CD & SEX & NATIONALITY CD & BIRTH DATE & OCCUPATION CODE & INCOME GROUP CD & MARRIED*/

cust_ind_1 = cust_ind.drop(columns=['AS_OF_DATE'])

CC04 = CC03.merge(cust_ind_1.rename(columns={'CUS_NO': 'M_Cus_No'}), on="M_Cus_No", how="left")

CC04['M_Product_Hierarchy_Cd'] = str(0) + CC04['M_Product_Hierarchy_Cd'].astype(str)
CC04.sort_values(by=['M_Product_Hierarchy_Cd'], ascending=True, inplace=True)

PRD1 = Product_Hierarchy[['Product_Hierarchy_Cd','Level_4','Level_5','Level_6']]

PRD1.Product_Hierarchy_Cd = PRD1.Product_Hierarchy_Cd.astype(str)
PRD1.Product_Hierarchy_Cd = PRD1.Product_Hierarchy_Cd.str.upper()
PRD1.Product_Hierarchy_Cd = PRD1.Product_Hierarchy_Cd.str.strip()

CC05 = CC04.merge(PRD1.rename(columns={'Product_Hierarchy_Cd':'M_Product_Hierarchy_Cd'}), on="M_Product_Hierarchy_Cd", how="left")


#SOURCE PROD TYPE DESC

CC05['M_Source_Prod_Type_Cd'] = str(0) + CC05['M_Source_Prod_Type_Cd'].astype(str)
CC05.sort_values(by=['M_Source_Prod_Type_Cd'], ascending=True, inplace=True)

PT002 = Product_Hierarchy.rename(columns={'Source_Prod_Type_Cd':'M_Source_Prod_Type_Cd'})[['M_Source_Prod_Type_Cd','Source_Prod_Type_Dsc']]

CC06 = CC05.merge(PT002, on="M_Source_Prod_Type_Cd", how="left")

CC06.sort_values(by=['M_INCOME_GRP_CD'], ascending=True, inplace=True)

#GCIF INCOME

INCOM=Income_Grp_Dsc.rename(columns={'Income_Grp_Cd':'M_INCOME_GRP_CD'})[['Income_Grp','M_INCOME_GRP_CD']]

CC07 = pd.merge(CC06, INCOM, on="M_INCOME_GRP_CD", how="left")


def Race(M_RACE_CD):
    if M_RACE_CD==1:
        return 'Malay'
    elif M_RACE_CD==2:
        return 'Native'
    elif M_RACE_CD==3:
        return 'Chinese'
    elif M_RACE_CD==4:
        return 'Indian'
    elif M_RACE_CD==99:
        return 'Others'
    else: 
        return 'Not Populated' 
CC07['RACE'] = CC07['M_RACE_CD'].apply(Race)


def Gender(SEX):
    if SEX==1:
        return 'Female'
    elif SEX==2: 
        return 'Male'
    else: 
        return 'Not Populated'
CC07['GENDER'] = CC07['SEX'].apply(Gender)


CC07.BTH_DT = CC07.BTH_DT.astype(str)
CC07['BIRTHDAYYEAR'] = CC07.BTH_DT.str[-6:]

CC07.BIRTHDAYYEAR = CC07.BIRTHDAYYEAR.astype(float)

def AGE(BIRTHDAYYEAR):
    if BIRTHDAYYEAR == 0:
        return 0
    else:
        return 2022 - BIRTHDAYYEAR
CC07['AGE'] = CC07['BIRTHDAYYEAR'].apply(AGE)


def AGE_RANGE(AGE):
    if AGE >= 1 and AGE < 25: 
        return 'A. <25'
    elif AGE >= 25 and AGE < 30:
        return 'B. 25-29'
    elif AGE >= 30 and AGE < 35:
        return 'C. 30-34'
    elif AGE >= 35 and AGE < 40: 
        return 'D. 35-39'
    elif AGE >= 40 and AGE < 45: 
        return 'E. 40-44'
    elif AGE >= 45 and AGE < 50: 
        return 'F. 45-49'
    elif AGE >= 50 and AGE < 55: 
        return 'G. 50-54'
    elif AGE >= 55 and AGE < 99: 
        return 'H. >=55'
    else: 
        return 'I. NOT POPULATED'
CC07['AGE_RANGE'] = CC07['AGE'].apply(AGE_RANGE)


def MARITAL_STATUS(MARRIED):
    if MARRIED=='1': 
        return 'Single'
    elif MARRIED=='2':
        return 'Married'
    else: 
        return 'Not Populated'
CC07['MARITAL_STATUS'] = CC07['MARRIED'].apply(MARITAL_STATUS)


CC07 = CC07.drop(['BIRTHDAYYEAR'], axis = 1)


CC07_1 = CC07.fillna(0)

CC07_1.loc[(CC07_1.M_Doubtful_Flag=='N')&(CC07_1.M_Mths_In_Arrears.isin([2,3])),'F_SMA']='Y'
CC07_1.F_SMA.fillna('N',inplace=True)

CC07_1.rename(columns={'M_Doubtful_Flag':'M_NPL_FLG'}, inplace=True)

CC07_1['Yr_Open'] = CC07_1.M_Acct_Open_Date.str[-4:]

CC07_1['NOA'] = 1

SUMM_CC = CC07_1.groupby(['PD_SEGMENT'\
,'Income_Grp'\
,'RACE'\
,'Level_6'\
,'Source_Prod_Type_Dsc'\
,'M_Principal_Flg'\
,'M_NPL_FLG'\
,'F_SMA'\
,'Yr_Open'])[['M_Bnm_Balance2','Ead_Proportional_Myr','Expected_Loss_Myr','Rwa_After_Risk_Mit_Myr','NOA']].sum().reset_index()


SUMM_CC.rename(columns={'M_Bnm_Balance2':'M_Bnm_Balance',\
                        'Ead_Proportional_Myr':'EAD',\
                        'Expected_Loss_Myr':'EL',\
                        'Rwa_After_Risk_Mit_Myr':'RWA'}, inplace=True)
```

```python
CC07.to_csv(r'T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\12_CC\\'+str(date)+'\infocube_12_CC'+str(date)+'.txt', index = False)

SUMM_CC.to_excel(r'T:\MIB Risk Management\Credit Risk Analytics - Historical\04_Infocube\12_CC\\'+str(date)+'\\SUMM_CC'+str(date)+'(python).xlsx', sheet_name='SUMM', index = False)
```
