```python
#===========================================================================================================================

riskcatCBBB = merge_a.iloc[np.where(merge_a['COUNTRY/SECTOR'].isin(['GB','BB']))]



riskcatCBBB.loc[riskcatCBBB['BRR(CURRENT)']=="Unrated",'BRR(CURRENT)_2'] = -9999
riskcatCBBB.loc[riskcatCBBB['BRR(CURRENT)']!="Unrated",'BRR(CURRENT)_2'] = riskcatCBBB['BRR(CURRENT)']
riskcatCBBB.loc[riskcatCBBB['BRR(CURRENT)']=="NA",'BRR(CURRENT)_2'] = -9999
riskcatCBBB.loc[riskcatCBBB['BRR(CURRENT)']=="PF (2-GD)",'BRR(CURRENT)_2'] = -9999 #case jun 21

#riskcatCBBB['BRR(CURRENT)_2'] = riskcatCBBB['BRR(CURRENT)'].str.strip()
riskcatCBBB['BRR(CURRENT)_2'].fillna(-9999,inplace=True)
riskcatCBBB['BRR(CURRENT)_2'] = riskcatCBBB['BRR(CURRENT)_2'].astype(int)


riskcatCBBB.GCIF = riskcatCBBB.GCIF.astype(str)
riskcatCBBB.GCIF = riskcatCBBB.GCIF.str.strip()
riskcatCBBB.GCIF = riskcatCBBB.GCIF.str.upper()

CB.M_CUS_NO = CB.M_CUS_NO.astype(str)
CB.M_CUS_NO = CB.M_CUS_NO.str.strip()
CB.M_CUS_NO = CB.M_CUS_NO.str.upper()

CB1 = CB.iloc[np.where(CB.SEGMENT!='03. SME')][['M_CUS_NO','FINAL_RATING']].rename(columns={'M_CUS_NO':'GCIF'})

riskcatCBBB = riskcatCBBB.merge(CB1, on='GCIF', how='left', indicator='exist')

riskcatCBBB.loc[(riskcatCBBB.exist=='both')&(riskcatCBBB['BRR(CURRENT)_2']==-9999),'BRR(CURRENT)_2'] = riskcatCBBB.FINAL_RATING

riskcatCBBB['BRR(CURRENT)_2'] = riskcatCBBB['BRR(CURRENT)_2'].astype(int)

#a.iloc[np.where((a.exist=='both')&(a['BRR(CURRENT)']==-9999))] cek list yg wong xde tp kita ad rating


def RISKCD(x,y):
    if x >= 1 and x<= 5:
        return '01.Very Low Risk'
    elif y == 'Good / Normal':
        return '01.Very Low Risk'
    elif x >= 6 and x <= 10:
        return '02.Low Risk'
    elif y == 'EWS' or y == 'NCEWS':
        return '02.Low Risk'
    elif x >= 11 and x <= 15:
        return '03.Moderate Risk'
    elif y =='WL' or y == 'Watchlist'or y == 'NCWL':
        return '03.Moderate Risk'
    elif x >= 16 and x <= 21:
        return '04.High Risk'
    elif y=='SMA':
        return '04.High Risk'
    elif x >= 22 : #and x <= 23
        return '05.Impaired' #impaired '05.Sub Standard'
    elif y=='Impaired Non-Performing Loans'or y == 'Impaired Performing Loans'or y == 'GIL'or y == 'NCIL':
        return '05.Impaired'
    else:
        return '06.Unrated'
    
#riskcatCBBB['RISK_CATEGORY'] = riskcatCBBB['BRR(CURRENT)_2'].apply(RISKCD)

riskcatCBBB['RISK_CATEGORY'] = riskcatCBBB.apply(lambda x: RISKCD(x['BRR(CURRENT)_2'],x['STATUS_OF_ACCOUNT(CURRENT)']), axis=1)

```
