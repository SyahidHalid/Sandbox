```python
#AITAB
RDATEC = 202206 #/*current quarter*/
RDATEP =  202112 #/*last quarter*/

#Export
RDATE2 = '2022-06' #/*current quarter*/

#MOB
RATE_C = '30/06/2022' 
RATE_P = '31/12/2021'

#HOME_FINANCING
#RDATE3C = 20211231 #/*current quarter*/
#RDATE3P = 20210630 #/*last quarter*/

import numpy as np
import pandas as pd
import pandasql
import sqlite3

pd.set_option("display.max_columns", None) 
pd.set_option("display.max_colwidth", 1000) #huruf dlm column
pd.set_option("display.max_rows", 100)
pd.set_option("display.precision", 2) #2 titik perpuluhan


AITAB = pd.read_csv(r"T:\MIB Risk Management\Credit Risk Analytics - Historical\02_Data_Source\04_AITAB_Financing\04_Dataset\\"+str(RDATEC)+"\HP_"+str(RDATEC)+".txt", sep = "\t", header = 0, error_bad_lines = False, encoding='cp1252')
AITAB_P = pd.read_csv(r"T:\MIB Risk Management\Credit Risk Analytics - Historical\02_Data_Source\04_AITAB_Financing\04_Dataset\\"+str(RDATEP)+"\HP_"+str(RDATEP)+".txt", sep = "\t", header = 0, error_bad_lines = False, encoding='cp1252')

AUTO = pd.read_excel (r"T:\MIB Risk Management\Credit Risk Analytics - Historical\02_Data_Source\04_AITAB_Financing\04_Dataset\\"+str(RDATEC)+"\Auto_NCIF.xlsx", sheet_name='Sheet1')
AUTO_P = pd.read_excel (r"T:\MIB Risk Management\Credit Risk Analytics - Historical\02_Data_Source\04_AITAB_Financing\04_Dataset\\"+str(RDATEP)+"\Auto_NCIF.xlsx", sheet_name='Sheet1')


AUTO.columns = AUTO.columns.str.replace(" ", "_")
AUTO.rename(columns = {'Ac_No':'Account_No'}, inplace = True)

AUTO_P.columns = AUTO_P.columns.str.replace(" ", "_")
AUTO_P.rename(columns = {'Ac_No':'Account_No'}, inplace = True)

AUTO.Account_No = AUTO.Account_No.astype(str)
AUTO.Account_No = AUTO.Account_No.str.upper()
AUTO.Account_No = AUTO.Account_No.str.strip()

AUTO_P.Account_No = AUTO_P.Account_No.astype(str)
AUTO_P.Account_No = AUTO_P.Account_No.str.upper()
AUTO_P.Account_No = AUTO_P.Account_No.str.strip()

AITAB.sort_values(by=["Account_No"],ascending=True, inplace=True)
AITAB.Account_No = AITAB.Account_No.astype(str)
AITAB.Account_No = AITAB.Account_No.str.upper()
AITAB.Account_No = AITAB.Account_No.str.strip()

AITAB_P.sort_values(by=["Account_No"],ascending=True, inplace=True)
AITAB_P.Account_No = AITAB_P.Account_No.astype(str)
AITAB_P.Account_No = AITAB_P.Account_No.str.upper()
AITAB_P.Account_No = AITAB_P.Account_No.str.strip()

AITAB1 = AITAB.merge(AUTO, on="Account_No", how = 'left', indicator = True)
AITAB_P1 = AITAB_P.merge(AUTO_P, on="Account_No", how = 'left', indicator = True)
```

```python
def FOUND(_merge):
    if _merge == 'both':
        return 'Y'
    else:
        return 'N'
AITAB1['FOUND'] = AITAB1._merge.apply(FOUND)
AITAB_P1['FOUND'] = AITAB_P1._merge.apply(FOUND)

AITAB2 = AITAB1.drop(columns=["_merge"])
AITAB_P2 = AITAB_P1.drop(columns=["_merge"])

#/******************************* AITAB(HP01) *******************************/

HP01 = AITAB2[['As_Of_Date','Account_No','RISK_GRADE_VAL','NPL_Flag','FACILITY_TYPE','Cus_No','BNM_BALANCE']]

#HP01.BNM_BALANCE.fillna(0,inplace=True)

HP01_1 = HP01.iloc[np.where((HP01.FACILITY_TYPE=='I')&(HP01.BNM_BALANCE!=0))]

#/******************************* AITAB(HP02) *******************************/

HP02 = AITAB_P2[['As_Of_Date','Account_No','RISK_GRADE_VAL','NPL_Flag','FACILITY_TYPE','Cus_No','BNM_BALANCE']]

HP02_1 = HP02.iloc[np.where(HP02.FACILITY_TYPE=='I')]

#/******************************* AITAB(HP03) *******************************/

HP02_1['Account_No_P'] = HP02_1['Account_No']

HP02_1.rename(columns = {'RISK_GRADE_VAL':'RISK_GRADE_VAL_P','NPL_Flag':'NPL_Flag_P'}, inplace = True)

HP02_1.sort_values(by=["Account_No"],ascending=True, inplace=True)

HP03 = HP01_1.merge(HP02_1[['Account_No','RISK_GRADE_VAL_P','NPL_Flag_P','Account_No_P']], on="Account_No", how = 'left')
```

```python
#/******************************* AITAB(HP04) *******************************/

HP03.NPL_Flag = HP03.NPL_Flag.str.strip()

def PD_Grade2(NPL_Flag,RISK_GRADE_VAL):
    if NPL_Flag == 'Y':
        return 'IF'
    else:
        return RISK_GRADE_VAL
HP03['PD_Grade2'] = HP03.apply(lambda x: PD_Grade2(x['NPL_Flag'], x['RISK_GRADE_VAL']), axis=1)


HP03.PD_Grade2 = HP03.PD_Grade2.str.strip()

def RISK_CAT(PD_Grade2):
    if PD_Grade2 == 'R1':
        return 'VERY LOW'
    elif PD_Grade2 == 'R2':
        return 'VERY LOW'
    elif PD_Grade2 == 'R3':
        return 'LOW'
    elif PD_Grade2 == 'R4':
        return 'LOW'
    elif PD_Grade2 == 'R5':
        return 'MODERATE'
    elif PD_Grade2 == 'R6':
        return 'MODERATE'
    elif PD_Grade2 == 'R7':
        return 'MODERATE'
    elif PD_Grade2 == 'R8':
        return 'MODERATE'
    elif PD_Grade2 == 'R9':
        return 'HIGH'
    elif PD_Grade2 == 'R10':
        return 'HIGH'
    elif PD_Grade2 == 'R11':
        return 'HIGH'
    else:
        return PD_Grade2
HP03['RISK_CAT'] = HP03.apply(lambda x: RISK_CAT(x['PD_Grade2']), axis=1)


HP03.NPL_Flag_P = HP03.NPL_Flag_P.str.strip()

def PD_Grade_P2(NPL_Flag_P,RISK_GRADE_VAL_P):
    if NPL_Flag_P == 'Y':
        return 'IF'
    else:
        return RISK_GRADE_VAL_P
HP03['PD_Grade_P2'] = HP03.apply(lambda x: PD_Grade_P2(x['NPL_Flag_P'], x['RISK_GRADE_VAL_P']), axis=1)

HP03.PD_Grade_P2 = HP03.PD_Grade_P2.str.strip()

HP03.Account_No_P.fillna(-9999, inplace=True)

def RISK_CAT_P(PD_Grade_P2,Account_No_P):
    if PD_Grade_P2 == 'R1':
        return 'VERY LOW'
    elif PD_Grade_P2 == 'R2':
        return 'VERY LOW'
    elif PD_Grade_P2 == 'R3':
        return 'LOW'
    elif PD_Grade_P2 == 'R4':
        return 'LOW'
    elif PD_Grade_P2 == 'R5':
        return 'MODERATE'
    elif PD_Grade_P2 == 'R6':
        return 'MODERATE'
    elif PD_Grade_P2 == 'R7':
        return 'MODERATE'
    elif PD_Grade_P2 == 'R8':
        return 'MODERATE'
    elif PD_Grade_P2 == 'R9':
        return 'HIGH'
    elif PD_Grade_P2 == 'R10':
        return 'HIGH'
    elif PD_Grade_P2 == 'R11':
        return 'HIGH'
    elif Account_No_P == -9999:
        return 'NEW'
    else:
        return PD_Grade_P2
HP03['RISK_CAT_P'] = HP03.apply(lambda x: RISK_CAT_P(x['PD_Grade_P2'], x['Account_No_P']), axis=1)

HP03['NOA'] = 1


HP03.fillna(0,inplace=True)

HP04 = HP03.groupby(['RISK_GRADE_VAL','PD_Grade2','RISK_CAT','NPL_Flag','RISK_GRADE_VAL_P','PD_Grade_P2','RISK_CAT_P','NPL_Flag_P'])[['NOA','BNM_BALANCE']].sum().reset_index()

```
