```python
import pandas as pd
import numpy as np
```

```python
LDB = pd.read_excel(r"C:\Users\syahidhalid\Syahid_PC\Analytics - FAD\00. Automation LDB\Data Source\202406\Loan Database as Jun 2024 (python) - Final.xlsx", sheet_name="Export Worksheet", header=1)
```

```python
LDB1 = LDB.groupby(['Fund / Guarantee /Program / Insentif'])[['Customer Name']].count().reset_index()
```

```python
LDB1['Fund / Guarantee /Program / Insentif'] =  LDB1['Fund / Guarantee /Program / Insentif'].str.replace("\n"," ")
LDB1['Fund / Guarantee /Program / Insentif'] =  LDB1['Fund / Guarantee /Program / Insentif'].str.strip()
```

# Fund Type

```python
#LDB1['Fund / Guarantee /Program / Insentif'].value_counts()
```

```python
#LDB1.iloc[np.where(LDB1['Fund / Guarantee /Program / Insentif'].str.contains("BNM"))]

LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("BNM Fund - E-FIRST (All Economic Sectors)"), "Fund Type"] = "BNM Fund - E-FIRST (All Economic Sectors)"
LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("BNM Fund - E-FIRST (All Economic Sectors)"), "Fund Type"] = "BNM Fund - E-FIRST (All Economic Sectors)"
LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("BNM Fund - E-FIRST (Special Relief Facility)"), "Fund Type"] = "BNM Fund - E-FIRST (Special Relief Facility)"
LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("BNM Fund - High Tech & Green Facility"), "Fund Type"] = "BNM Fund - High Tech & Green Facility"
LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("Fund BNM - High Tech"), "Fund Type"] = "BNM Fund - High Tech"
LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("BNM Fund - All Economic Sector"), "Fund Type"] = "BNM Fund - E-FIRST (All Economic Sectors)"
```

# Insentif

```python
#LDB1.iloc[np.where((LDB1['Fund / Guarantee /Program / Insentif'].str.contains("Insentif"))
#                   |(LDB1['Fund / Guarantee /Program / Insentif'].str.contains("TERAJU")))]
LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("Insentif - Skim Insentif Pemampanan Pengeksport (SIP2)"), "Insentif"] = "Skim Insentif Pemampanan Pengeksport (SIP2)"
LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("TERAJU"), "Insentif"] = "TERAJU"
LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("Skim Insentif Pemampanan Pengeksport (SIP2)"), "Insentif"] = "Skim Insentif Pemampanan Pengeksport (SIP2)"
```

# Guarantee

```python
#LDB1.iloc[np.where(LDB1['Fund / Guarantee /Program / Insentif'].str.contains("SJPP","CGC"))]

LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("Guarantee - SJPP"), "Guarantee"] = "SJPP"
LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("Guarantee -  SJPP"), "Guarantee"] = "SJPP"
LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("SJPP"), "Guarantee"] = "SJPP"
LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("CGC"), "Guarantee"] = "CGC"
```

# Program

```python
#LDB1.iloc[np.where(LDB1['Fund / Guarantee /Program / Insentif'].str.contains("Program"))]

LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("Program  - SME Export"), "Programme"] = "SME Export"
LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("Programme SME-Export"), "Programme"] = "SME Export"
LDB1.loc[LDB1['Fund / Guarantee /Program / Insentif'].str.contains("Program -ACE (Anchor Company Export Ecosystem)"), "Programme"] = "ACE (Anchor Company Export Ecosystem)"
#Green Tech Export
```

```python
LDB1['Fund Type'] = LDB1['Fund Type'].str.replace("nan","")
LDB1['Insentif'] = LDB1['Insentif'].str.replace("nan","")
LDB1['Guarantee'] = LDB1['Guarantee'].str.replace("nan","")
LDB1['Programme'] = LDB1['Programme'].str.replace("nan","")
```

```python
LDB1.to_excel(r"C:\Users\syahidhalid\Syahid_PC\Analytics - FAD\00. Automation LDB\Data Source\202406\Fund Type - Data Dictionary.xlsx", sheet_name="Export Worksheet",index=False)
```
