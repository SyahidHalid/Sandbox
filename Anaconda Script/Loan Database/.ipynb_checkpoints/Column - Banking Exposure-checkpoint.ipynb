```python

LDB = pd.read_excel(str(Location)+"\\Loan Database as Jul 2024 (python) - Final.xlsx", sheet_name='Export Worksheet', header=1) # skiprows = 2, nrows=18,  usecols= 'A:D'
```

```python
LDB1 = LDB.iloc[np.where(LDB['EXIM Account No.']!="Total")]
```

```python
LDB1.head(1)
```

```python
def NOB(a_exp,b_exp,c_exp,d_exp,f_exp):
    if (a_exp=='Non Trade')&((b_exp=="No Further Disbursement")|(b_exp=="Fully Disbursed")):
        return d_exp
    elif (a_exp=='Non Trade')&((b_exp=="Ongoing Disbursement")|(b_exp=="Pending Disbursement")): 
        return f_exp
    elif ((a_exp=='Trade')|(a_exp=='Trade - Guarantee'))&((b_exp=="Ongoing Disbursement")|(b_exp=="Pending Disbursement")): 
        return f_exp
    elif ((a_exp=='Trade')|(a_exp=='Trade - Guarantee'))&((b_exp=="No Further Disbursement")|(b_exp=="Fully Disbursed")): 
        return d_exp
    elif c_exp == 'Impaired': 
        return d_exp
    else: 
        return 0
LDB1['TBE (Facility Currency)'] = LDB1.apply(lambda x: NOB(x['Nature of Account'],x['Disbursement/\nDrawdown Status'],x['Status'],x['Cost/Principal Outstanding (Facility Currency)'],x['Amount Approved / Facility Limit (Facility Currency)']), axis=1)

LDB1['TBE (MYR)'] = LDB1.apply(lambda x: NOB(x['Nature of Account'], x['Disbursement/\nDrawdown Status'], x['Status'], x['Cost/Principal Outstanding (MYR)'],x['Amount Approved / Facility Limit (MYR)']), axis=1)
```
